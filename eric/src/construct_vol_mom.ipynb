```python
import cPickle as pickle
### NOW WANT TO CONSTRUCT VOLATILITY AND MOMENTUM MEASURES.
price_data_full
ewma_30d = pd.stats.moments.ewma(price_data_full, com=14.5)
ewma_15d = pd.stats.moments.ewma(price_data_full, com=7.0)
ewma_5d = pd.stats.moments.ewma(price_data_full, com=2.0)

def get_mom(smoothed_price,lag):
    '''GET THE DIFFERENCE IN THE SMOOTHED MOVING AVERAGE BETWEEN PREVIOUS TRADING DAY AND lag PERIODS AGO, 
    OR THE FEWEST NUMBER OF DAYS STRICTLY GREATER THAN lag FOR WHICH PRICE DATA IS AVAILABLE'''
    mom = pd.DataFrame(data=None,index=smoothed_price.index,columns=smoothed_price.columns)
    for dates in smoothed_price.index:
        counter=0
        while True:
            if min(smoothed_price.index)>(dates - timedelta64(lag+counter,'D')):               
                break
            close_ndays = lag+counter
            date_match = dates-timedelta64(close_ndays,'D')
            if any(smoothed_price.index==date_match):
                mom.loc[dates]= 100.0*365.0/close_ndays*(smoothed_price.loc[dates]-smoothed_price.loc[date_match])/smoothed_price.loc[date_match]               
                break
            else:
                counter+=1           
    return mom

def get_fwd_vol(price,num_days):
    ### GET VOLATILITY COMPUTED FROM THE CURRENT DATE TO CURRENT DATE + num_days IN THE FUTURE
    returns = price.pct_change()
    fwd_vol = pd.DataFrame(data=None,index=price.index,columns=price.columns)
    for dates in price.index:
        fwd_flag=dates+timedelta64(num_days,'D')
        good_index = logical_and((returns.index<=fwd_flag), (returns.index>=dates))
        fwd_vol.loc[dates] = std(100.0*(252.0**0.5)*returns.loc[returns.index[good_index]])
    return fwd_vol

def get_back_vol(price,num_days):
    ### GET VOLATILITY COMPUTED FROM THE CURRENT DATE TO CURRENT DATE - 1 TO CURRENT DATE-1-num_days IN THE PAST
    returns = price.pct_change()
    back_vol = pd.DataFrame(data=None,index=price.index,columns=price.columns)
    for dates in price.index:
        back_flag=dates-timedelta64(num_days,'D')
        good_index = logical_and((returns.index>=back_flag), (returns.index<dates))
        back_vol.loc[dates] = std(100.0*(252.0**0.5)*returns.loc[returns.index[good_index]])
    return back_vol

fwd_vol = {}
num_days_fwd_vol = [5,10,15,20,25,30,60,90,120,150,180,210,250]
k=0
while k<len(num_days_fwd_vol):
    fwd_vol[num_days_fwd_vol[k]] = get_fwd_vol(price_data_full,num_days_fwd_vol[k])
    k+=1
    
num_days_back_vol = [2,3,4,5,6,7,15,30,50,75,100,125,175,200,225,250]  
num_days_mom = [1,2,3,4,5,6,7,15,30,50,75,100,125,175,200,225,250]    
    
back_vol = {}
k=0
while k<len(back_vol):
    back_vol[num_days_back_vol[k]] = get_back_vol(price_data_full,num_days_back_vol[k])
    k+=1
    
mom_5dewma = {}
mom_15dewma = {}
mom_30dewma = {}
abs_mom_5dewma = {}
abs_mom_15dewma = {}
abs_mom_30dewma = {}  

k=0   
while k<len(num_days_mom):
    mom_5dewma[num_days_mom[k]] = get_mom(ewma_5d,num_days_mom[k])
    mom_15dewma[num_days_mom[k]] = get_mom(ewma_15d,num_days_mom[k])
    mom_30dewma[num_days_mom[k]] = get_mom(ewma_30d,num_days_mom[k])
    
    abs_mom_5dewma[num_days_mom[k]] = abs(mom_5dewma[num_days_mom[k]])
    abs_mom_15dewma[num_days_mom[k]] = abs(mom_15dewma[num_days_mom[k]])
    abs_mom_30dewma[num_days_mom[k]] = abs(mom_30dewma[num_days_mom[k]])
    k+=1

data_list = [fwd_vol, back_vol,mom_5dewma,mom_15dewma,mom_30dewma, abs_mom_5dewma,abs_mom_15dewma,abs_mom_30dewma]
data_list_save = ["fwd_vol_save.dat","back_vol_save.dat","mom_5dewma_save.dat","mom_15dewma_save.dat","mom_30dewma_save.dat",\
                  "abs_mom_5dewma_save.dat","abs_mom_15dewma_save.dat","abs_mom_30dewma_save.dat"]

for items,save_items in zip(data_list,data_list_save):
    with open(save_items, "wb") as f:
        pickle.dump(items, f)

```

```python
### LATER RUN WITH SOME COMBINATION OF MORE STOCKS, WITH VALUE INDICATORS/INDUSTRY INDICATORS, BETA, SIZE, ANY OTHER INDICATORS
from sklearn.linear_model import Ridge, Lasso
from sklearn import linear_model
import sklearn
import numpy as np
from matplotlib import pylab as plt
import Quandl

'''
INSERT QUANDL TOKEN HERE
'''
qdltoken = "" 

def make_mat_yvar(dep_var):
    '''TAKES A DATA FRAME, AND CONSTRUCTS A MATRIX OF APPROPRIATE SIZE 
    FOR THE DEPENDENT VARIABLE FOR INPUT INTO REGRESSION EQUATION'''
    dep_var_values = dep_var.values
    dep_var_mat = transpose(dep_var_values).flatten().reshape(dep_var_values.size,1)
    return dep_var_mat

def clean_reg_matr(X,y):
    '''REMOVES INSTANCES OF NAN FROM THE REGRESSION MATRICES FOR MATRICES'''
    con_mat = np.concatenate((X,y),1)
    good_indices = logical_not(any(np.isnan(con_mat.astype(float)),1)).astype(bool)
    X_trimmed = X[good_indices]
    y_trimmed = y[good_indices]
    return X_trimmed, y_trimmed
    

def make_mat_xvars(*args):
    '''TAKES A DATA FRAME, AND CONSTRUCTS A MATRIX OF APPROPRIATE SIZE 
    FOR THE DEPENDENT VARIABLE FOR INPUT INTO REGRESSION EQUATION'''
    ### GET THE DIMENSIONS
    num_dep_vars = args[0].shape[1]
    num_obs = args[0].shape[0]
    num_vars = len(args)
    ### INITIALIZE THE MATRIX    
    ind_var_mat = np.zeros((num_obs*num_dep_vars,num_vars+num_dep_vars-1)) 
    ### MAKE SURE THE NAMES IN THE VARIOUS DATAFRAMES ALIGN - IF NOT RETURN AN ERROR
    for check_var_counter in range(num_vars-1):
        if not all(args[check_var_counter].columns==args[check_var_counter+1].columns):
            print "Mismatched Data Columns In Constructing Regression Matrices"
            return None

    ### MAKE SURE ALL THE DATES ALIGN FOR THE DEPENDENT VARIABLES
    for xvar_counter in range(num_vars-1):
        if not all(args[xvar_counter].index == args[xvar_counter+1].index):
            print "Row indices (Dates) are misaligned"
            return None
        
    counter_dim1=0
    dummy_counter=0
    for var_names in args[0].columns:
        counter_dim2=0
        for index in range(num_vars):
            ind_var_mat[counter_dim1:counter_dim1+num_obs,counter_dim2] = args[index][var_names].values 
            counter_dim2+=1
        if dummy_counter>0:
            ind_var_mat[counter_dim1:counter_dim1+num_obs,counter_dim2-1+dummy_counter] = 1.0
        dummy_counter += 1
        counter_dim1+=num_obs     
    return ind_var_mat

def make_mat_xvars_civ(common_independent_vars,*args):
    '''TAKES A DATA FRAME, AND CONSTRUCTS A MATRIX OF APPROPRIATE SIZE 
    FOR THE DEPENDENT VARIABLE FOR INPUT INTO REGRESSION EQUATION
    GET THE DIMENSIONS.  RELATIVE TO make_mat_xvars, THIS FUNCTION
    ALLOWS FOR US TO PASS INDEPENDENT VARIABLES WHICH ARE THE SAME ACROSS
    DIFFERENT VARIABLES, BUT THE SAME AT EACH TIME POINT FOR EVERY STOCK WE 
    ARE LOOKING AT (EG MACROECONOMIC/AGGREGATE FINANCIAL VARIABLES)'''
    num_dep_vars = args[0].shape[1]
    num_obs = args[0].shape[0]
    num_vars = len(args)
    num_civ = common_independent_vars.shape[1]
    names_civ = common_independent_vars.columns
    
    ### INITIALIZE THE MATRIX    
    ind_var_mat = np.zeros((num_obs*num_dep_vars,num_vars+num_dep_vars-1+num_civ))
    
    ### MAKE SURE THE NAMES IN THE VARIOUS DATAFRAMES ALIGN - IF NOT RETURN AN ERROR
    for check_var_counter in range(num_vars-1):
        if not all(args[check_var_counter].columns==args[check_var_counter+1].columns):
            print "Mismatched Data Columns In Constructing Regression Matrices"
            return None
        
    ### MAKE SURE ALL THE DATES ALIGN FOR THE DEPENDENT VARIABLES
    for xvar_counter in range(num_vars-1):   
        if not all(args[xvar_counter].index == args[xvar_counter+1].index):
            print "Row indices (Dates) are misaligned"     
            return None
    
    ### MAKE SURE THAT COMMON INDEPENDENT VARIABLE DATES ALIGN WITH THE OTHER INDEPENDENT VARIABLES
    if not all(args[0].index == common_independent_vars.index):
        print "Indices not aligned between the common independent variables and the other independent variables, merge first"
        return None
    
    counter_dim1=0
    dummy_counter=0
    for var_names in args[0].columns:
        counter_dim2=0
        index_civ=0
        for index in range(num_vars):
            ind_var_mat[counter_dim1:counter_dim1+num_obs,counter_dim2] = args[index][var_names].values 
            counter_dim2+=1
        for names in names_civ:
            ind_var_mat[counter_dim1:counter_dim1+num_obs,counter_dim2-1+index_civ] = common_independent_vars.ix[:,index_civ]
            index_civ+=1
        if dummy_counter>0:
            ind_var_mat[counter_dim1:counter_dim1+num_obs,counter_dim2+index_civ-2+dummy_counter] = 1.0
        dummy_counter += 1
        counter_dim1+=num_obs
        
    return ind_var_mat
    
def make_test_matrices(start_test_date,y_var,*args):
    ### MAKES THE SET OF MATRICES FOR THE TEST SET
    ### WANT TO BE ABLE TO ADJUST START MONTH, BUT MAKE SURE START YEAR IS ALWAYS 2009, OTHERWISE ERROR
    if start_test_date.year != 2009:
        print 'Construction of test matrices is incorrect - check dates'
        return None
    
    num_vars = len(args)
    y_mat = make_mat_yvar(y_var)
    X_mat = make_mat_xvars(*args)
    
    ### CHECK THAT THE TIME INDICES ALIGN, OTHERWISE WE ARE IN MAJOR TROUBLE
    num_vars = len(args)
    y_date = y_var.index
    for xvar_counter in range(num_vars):
        if not all(y_date==args[xvar_counter].index):
            print "Row indices (Dates) are misaligned"     
            return None
        
    ### NOW SPLIT THE X AND y MATRICES INTO 5 SETS TO USE FOR TESTING IN THE CROSS-VALIDATION
    con_mat = np.concatenate((X_mat,y_mat),1)
    good_indices = logical_not(any(np.isnan(con_mat.astype(float)),1)).astype(bool)
    y_date_rep = tile(y_date,(args[0].shape[1],1)).flatten('F') ### REPEAT THE DATE FOR NUMBER OF VARIABLES WE HAVE
    
    ytestmat = {}
    Xtestmat = {}
    
    dates_list = [2009,2010,2011,2012,2013]
    
    for dates in dates_list:
        end_date = start_test_date + timedelta64(12,'M')
        good_indices_test = logical_and(good_indices,logical_and(y_date_rep>start_test_date,y_date_rep<=end_date))
        ytestmat[dates] = y_mat[good_indices_test]
        Xtestmat[dates] = X_mat[good_indices_test] 
        start_test_date = end_date + timedelta64(1,'D') 
    
    return ytestmat, Xtestmat

def make_test_matrices_withciv(start_test_date,y_var,common_indvar,*args):
    ### MAKES THE SET OF MATRICES FOR THE TEST SET
    ### WANT TO BE ABLE TO ADJUST START MONTH, BUT MAKE SURE START YEAR IS ALWAYS 2009, OTHERWISE ERROR
    if start_test_date.year != 2009:
        print 'Construction of test matrices is incorrect - check dates'
        return None
    
    num_vars = len(args)
    y_mat = make_mat_yvar(y_var)
    X_mat = make_mat_xvars_civ(common_indvar,*args)
    
    ### CHECK THAT THE TIME INDICES ALIGN
    num_vars = len(args)
    y_date = y_var.index
    for xvar_counter in range(num_vars):
        if not all(y_date==args[xvar_counter].index):
            print "Row indices (Dates) are misaligned"     
            return None
        
    ### NOW SPLIT THE X AND y MATRICES INTO 5 SETS TO USE FOR TESTING IN THE CROSS-VALIDATION
    con_mat = np.concatenate((X_mat,y_mat),1)
    good_indices = logical_not(any(np.isnan(con_mat.astype(float)),1)).astype(bool)
    y_date_rep = tile(y_date,(args[0].shape[1],1)).flatten('F') ### REPEAT THE DATE FOR NUMBER OF VARIABLES WE HAVE
    
    ytestmat = {}
    Xtestmat = {}
    
    dates_list = [2009,2010,2011,2012,2013]
    
    for dates in dates_list:
        end_date = start_test_date + timedelta64(12,'M')
        good_indices_test = logical_and(good_indices,logical_and(y_date_rep>start_test_date,y_date_rep<=end_date))
        ytestmat[dates] = y_mat[good_indices_test]
        Xtestmat[dates] = X_mat[good_indices_test] 
        start_test_date = end_date + timedelta64(1,'D') 
    
    return ytestmat, Xtestmat

def make_train_valid_matrices_withcommonvars(start_train_date,end_train_date,start_valid_date,end_valid_date,y_var,common_ind_vars,*args):
    '''MAKES THE SET OF MATRICES FOR THE TRAINING AND VALIDATION SETS
    THIS FUNCTION IS LIKE MAKE_TRAIN_VALID_MATRICES, EXCEPT IT ALSO ALLOWS US TO INCLUDE VARIABLES WHICH ARE COMMON ACROSS
    ALL VARIABLES - LIKE MACROECONOMICS VARIABLES (VIX INDEX, STOCK RETURNS, VIX FUTURES, ETC.)'''
    y_mat = make_mat_yvar(y_var)
    X_mat = make_mat_xvars_civ(common_ind_vars,*args)
    
    ### CHECK THAT THE TIME INDICES ALIGN
    num_vars = len(args)
    y_date = y_var.index
    for xvar_counter in range(num_vars):
        if not all(y_date==args[xvar_counter].index):
            print "Row indices (Dates) are misaligned"     
            return None
    
    ### NOW SPLIT THE X AND y MATRICES INTO 2 DIFFERENT SETS TO USE AS INPUTS FOR THE CROSS-VALIDATION
    con_mat = np.concatenate((X_mat,y_mat),1)
    good_indices = logical_not(any(np.isnan(con_mat.astype(float)),1)).astype(bool)
    y_date_rep = tile(y_date,(args[0].shape[1],1)).flatten('F') ### REPEAT THE DATE FOR NUMBER OF VARIABLES WE HAVE
    good_indices_train = logical_and(good_indices,logical_and(y_date_rep>=start_train_date,y_date_rep<end_train_date))
    good_indices_valid = logical_and(good_indices,logical_and(y_date_rep>=start_valid_date,y_date_rep<end_valid_date))
    Xmat_train = X_mat[good_indices_train]
    ymat_train = y_mat[good_indices_train]  
    Xmat_valid = X_mat[good_indices_valid]
    ymat_valid = y_mat[good_indices_valid]   
    
    return Xmat_train, ymat_train, Xmat_valid, ymat_valid

def make_train_valid_matrices(start_train_date,end_train_date,start_valid_date,end_valid_date,y_var,*args):
    ### MAKES THE SET OF MATRICES FOR THE TRAINING AND VALIDATION SETS
    y_mat = make_mat_yvar(y_var)
    X_mat = make_mat_xvars(*args)
    
    ### CHECK THAT THE TIME INDICES ALIGN, OTHERWISE WE ARE IN MAJOR TROUBLE
    num_vars = len(args)
    y_date = y_var.index
    for xvar_counter in range(num_vars):
        if not all(y_date==args[xvar_counter].index):
            print "Row indices (Dates) are misaligned"     
            return None
    
    ### NOW SPLIT THE X AND y MATRICES INTO 2 DIFFERENT SETS TO USE AS INPUTS FOR THE CROSS-VALIDATION
    con_mat = np.concatenate((X_mat,y_mat),1)
    good_indices = logical_not(any(np.isnan(con_mat.astype(float)),1)).astype(bool)
    y_date_rep = tile(y_date,(args[0].shape[1],1)).flatten('F') ### REPEAT THE DATE FOR NUMBER OF VARIABLES WE HAVE
    good_indices_train = logical_and(good_indices,logical_and(y_date_rep>=start_train_date,y_date_rep<end_train_date))
    good_indices_valid = logical_and(good_indices,logical_and(y_date_rep>=start_valid_date,y_date_rep<end_valid_date))
    Xmat_train = X_mat[good_indices_train]
    ymat_train = y_mat[good_indices_train]  
    Xmat_valid = X_mat[good_indices_valid]
    ymat_valid = y_mat[good_indices_valid]   
    
    return Xmat_train, ymat_train, Xmat_valid, ymat_valid

    
### NEED TO MAKE SURE THESE GRIDS COVER APPROPRIATE SPACE - DO NOT WANT TO HAVE A CORNER SOLUTION HERE
ridge_grid = np.logspace(-5., 0.1, 20)
lasso_grid = np.logspace(-5., 0.1, 20)

### GET THE VIX AND S+P 500 DATA TO IMPROVE THE FORECASTS
### SINCE PRESUMABLY LEVEL OF VOLATILITY MATTERS A GREAT DEAL
### THINK ABOUT OTHER TRANSFORMATIONS/INDICATORS THAT COULD BE USEFUL - MARKET RETURN, TED SPREAD, VIX FUTURES, E.G.
vix_data = Quandl.get(['YAHOO/INDEX_VIX.6'],authtoken=qdltoken)
sp500_data = Quandl.get(['YAHOO/INDEX_GSPC.6'],authtoken=qdltoken)

civ_data = vix_data  ### LATER MAY WANT TO MERGE AND ADD OTHER AGGREGATE ECONOMIC/FINANCIAL VARIABLES - VIX CHANGE, CHANGE IN SP500, ETC.

civ_data_dates_big = civ_data.merge(fwd_vol_30d,how='right',right_index=True,left_index=True)
if civ_data.shape[1]>1:
    civ_data_dates_aligned = civ_data_dates_big.iloc[:,0:civ_data_dates_big.shape[1]]
else:
    civ_data_dates_aligned_series = civ_data_dates_big.iloc[:,0]
    civ_data_dates_aligned = pd.DataFrame(civ_data_dates_aligned)
    
num_predict_civ = civ_data_dates_aligned.shape[1]

num_test_sets = 5     
ytm_910, ytm_1011, ytm_1112, ytm_1213, ytm_1314, Xtm_910, Xtm_1011, Xtm_1112, Xtm_1213, Xtm_1314 = make_test_matrices(pd.to_datetime('2009-05-01'),fwd_vol_30d,mom_1d,mom_2d,mom_3d,mom_4d,mom_5d,mom_6d,mom_7d,mom_15d,mom_30d,mom_50d,mom_75d,mom_100d,mom_125d,mom_150d,mom_175d,mom_200d,mom_225d,mom_250d,abs_mom_1d,abs_mom_2d,abs_mom_3d,abs_mom_4d,abs_mom_5d,abs_mom_6d,abs_mom_7d,abs_mom_15d,abs_mom_30d,abs_mom_50d,abs_mom_75d,abs_mom_100d,abs_mom_125d,abs_mom_150d,abs_mom_175d,abs_mom_200d,abs_mom_225d,abs_mom_250d,back_vol_2d,back_vol_3d,back_vol_4d,back_vol_5d,back_vol_6d,back_vol_7d,back_vol_15d,back_vol_30d,back_vol_50d,back_vol_75d,back_vol_100d,back_vol_125d,back_vol_150d,back_vol_175d,back_vol_200d,back_vol_225d,back_vol_250d)
ytm_910_withciv, ytm_1011_withciv, ytm_1112_withciv, ytm_1213_withciv, ytm_1314_withciv, Xtm_910_withciv, Xtm_1011_withciv, Xtm_1112_withciv, Xtm_1213_withciv, Xtm_1314_withciv = make_test_matrices_withciv(pd.to_datetime('2009-05-01'),fwd_vol_30d,civ_data_dates_aligned,mom_1d,mom_2d,mom_3d,mom_4d,mom_5d,mom_6d,mom_7d,mom_15d,mom_30d,mom_50d,mom_75d,mom_100d,mom_125d,mom_150d,mom_175d,mom_200d,mom_225d,mom_250d,abs_mom_1d,abs_mom_2d,abs_mom_3d,abs_mom_4d,abs_mom_5d,abs_mom_6d,abs_mom_7d,abs_mom_15d,abs_mom_30d,abs_mom_50d,abs_mom_75d,abs_mom_100d,abs_mom_125d,abs_mom_150d,abs_mom_175d,abs_mom_200d,abs_mom_225d,abs_mom_250d,back_vol_2d,back_vol_3d,back_vol_4d,back_vol_5d,back_vol_6d,back_vol_7d,back_vol_15d,back_vol_30d,back_vol_50d,back_vol_75d,back_vol_100d,back_vol_125d,back_vol_150d,back_vol_175d,back_vol_200d,back_vol_225d,back_vol_250d)
n_predict = Xtm_1011.shape[1]
    
num_cv_tests = 92
lasso_rmse_store = np.zeros((num_cv_tests,num_test_sets))
ridge_rmse_store = np.zeros((num_cv_tests,num_test_sets))
lasso_mae_store = np.zeros((num_cv_tests,num_test_sets))
ridge_mae_store = np.zeros((num_cv_tests,num_test_sets))
lasso_r2_store = np.zeros((num_cv_tests,num_test_sets))
ridge_r2_store = np.zeros((num_cv_tests,num_test_sets))

lasso_rmse_store_withciv = np.zeros((num_cv_tests,num_test_sets))
ridge_rmse_store_withciv = np.zeros((num_cv_tests,num_test_sets))
lasso_mae_store_withciv = np.zeros((num_cv_tests,num_test_sets))
ridge_mae_store_withciv = np.zeros((num_cv_tests,num_test_sets))
lasso_r2_store_withciv = np.zeros((num_cv_tests,num_test_sets))
ridge_r2_store_withciv = np.zeros((num_cv_tests,num_test_sets))

naive_7_rmse_store = np.zeros((num_cv_tests,num_test_sets))
naive_15_rmse_store = np.zeros((num_cv_tests,num_test_sets))
naive_30_rmse_store = np.zeros((num_cv_tests,num_test_sets))
naive_50_rmse_store = np.zeros((num_cv_tests,num_test_sets))
naive_75_rmse_store = np.zeros((num_cv_tests,num_test_sets))

naive_7_mae_store = np.zeros((num_cv_tests,num_test_sets))
naive_15_mae_store = np.zeros((num_cv_tests,num_test_sets))
naive_30_mae_store = np.zeros((num_cv_tests,num_test_sets))
naive_50_mae_store = np.zeros((num_cv_tests,num_test_sets))
naive_75_mae_store = np.zeros((num_cv_tests,num_test_sets))

naive_7_r2_store = np.zeros((num_cv_tests,num_test_sets))
naive_15_r2_store = np.zeros((num_cv_tests,num_test_sets))
naive_30_r2_store = np.zeros((num_cv_tests,num_test_sets))
naive_50_r2_store = np.zeros((num_cv_tests,num_test_sets))
naive_75_r2_store = np.zeros((num_cv_tests,num_test_sets))

naive_7_r2_store_withciv = np.zeros((num_cv_tests,num_test_sets))
naive_15_r2_store_withciv = np.zeros((num_cv_tests,num_test_sets))
naive_30_r2_store_withciv = np.zeros((num_cv_tests,num_test_sets))
naive_50_r2_store_withciv = np.zeros((num_cv_tests,num_test_sets))
naive_75_r2_store_withciv = np.zeros((num_cv_tests,num_test_sets))

train_lasso_rmse_store = np.zeros((num_cv_tests,1))
train_ridge_rmse_store = np.zeros((num_cv_tests,1))
train_lasso_mae_store = np.zeros((num_cv_tests,1))
train_ridge_mae_store = np.zeros((num_cv_tests,1))
train_lasso_r2_store = np.zeros((num_cv_tests,1))
train_ridge_r2_store = np.zeros((num_cv_tests,1))

train_lasso_rmse_store_withciv = np.zeros((num_cv_tests,1))
train_ridge_rmse_store_withciv = np.zeros((num_cv_tests,1))
train_lasso_mae_store_withciv = np.zeros((num_cv_tests,1))
train_ridge_mae_store_withciv = np.zeros((num_cv_tests,1))
train_lasso_r2_store_withciv = np.zeros((num_cv_tests,1))
train_ridge_r2_store_withciv = np.zeros((num_cv_tests,1))

lasso_lambda_store = np.zeros((num_cv_tests,1))
ridge_lambda_store = np.zeros((num_cv_tests,1))
lasso_opt_coeff_store = np.zeros((num_cv_tests,n_predict))
ridge_opt_coeff_store = np.zeros((num_cv_tests,n_predict))

lasso_lambda_store_withciv = np.zeros((num_cv_tests,1))
ridge_lambda_store_withciv = np.zeros((num_cv_tests,1))
lasso_opt_coeff_store_withciv = np.zeros((num_cv_tests,n_predict+num_predict_civ))
ridge_opt_coeff_store_withciv = np.zeros((num_cv_tests,n_predict+num_predict_civ))

### CHECK WHETHER THE RELATIVE MOMENTUM, VOLATILITY ETC IS USEFUL FOR THE FORECASTING.

lasso_rmse_store_withrel = np.zeros((num_cv_tests,num_test_sets))
ridge_rmse_store_withrel = np.zeros((num_cv_tests,num_test_sets))
lasso_mae_store_withcrel = np.zeros((num_cv_tests,num_test_sets))
ridge_mae_store_withrel = np.zeros((num_cv_tests,num_test_sets))
lasso_r2_store_withcrel = np.zeros((num_cv_tests,num_test_sets))
ridge_r2_store_withcrel = np.zeros((num_cv_tests,num_test_sets))

train_lasso_rmse_store_withrel = np.zeros((num_cv_tests,1))
train_ridge_rmse_store_withrel = np.zeros((num_cv_tests,1))
train_lasso_mae_store_withrel = np.zeros((num_cv_tests,1))
train_ridge_mae_store_withrel = np.zeros((num_cv_tests,1))
train_lasso_r2_store_withrel = np.zeros((num_cv_tests,1))
train_ridge_r2_store_withrel = np.zeros((num_cv_tests,1))

lasso_lambda_store_withrel = np.zeros((num_cv_tests,1))
ridge_lambda_store_withrel = np.zeros((num_cv_tests,1))
lasso_opt_coeff_store_withrel = np.zeros((num_cv_tests,n_predict+num_predict_civ))
ridge_opt_coeff_store_withrel = np.zeros((num_cv_tests,n_predict+num_predict_civ))

### CONSTRUCT THE NORMALIZES VARIABLES TO SEE WHETHER RELATIVE MAGNITUDES MATTER MORE THAN ABSOLUTE LEVELS
### NEED TO FIGURE OUT WHAT TO SCALE THIS BY THOUGH SO THAT UNITS ARE APPROPRIATE
def standardize_data_frame(df,ax):
    stand_data = pd.dataframe(columns=df.columns, index = df.index, data=None)
    stand_data = (df - df.mean(axis=ax))/df.std(axis=ax)

### PREDICT VOLATILITY USING LAGGED VOLATILITY, SINCE VOLATILITY IS PROBABLY EXTREMELY PERSISTENT
rel_back_vol_2d = standardize_data_frame(back_vol_2d,1)
rel_back_vol_3d = standardize_data_frame(back_vol_3d,1)
rel_back_vol_4d = standardize_data_frame(back_vol_4d,1)
rel_back_vol_5d = standardize_data_frame(back_vol_5d,1)
rel_back_vol_6d = standardize_data_frame(back_vol_6d,1)
rel_back_vol_7d = standardize_data_frame(back_vol_7d,1)
rel_back_vol_15d = standardize_data_frame(back_vol_15d,1)
rel_back_vol_30d = standardize_data_frame(back_vol_30d,1)
rel_back_vol_50d = standardize_data_frame(back_vol_50d,1)
rel_back_vol_75d = standardize_data_frame(back_vol_75d,1)
rel_back_vol_100d = standardize_data_frame(back_vol_100d,1)
rel_back_vol_125d = standardize_data_frame(back_vol_125d,1)
rel_back_vol_150d = standardize_data_frame(back_vol_150d,1)
rel_back_vol_175d = standardize_data_frame(back_vol_175d,1)
rel_back_vol_200d = standardize_data_frame(back_vol_200d,1)
rel_back_vol_225d = standardize_data_frame(back_vol_225d,1)
rel_back_vol_250d = standardize_data_frame(back_vol_250d,1)

rel_mom_1d = standardize_data_frame(mom_1d,1)
rel_mom_2d = standardize_data_frame(mom_2d,1)
rel_mom_3d = standardize_data_frame(mom_3d,1)
rel_mom_4d = standardize_data_frame(mom_4d,1)
rel_mom_5d = standardize_data_frame(mom_5d,1)
rel_mom_6d = standardize_data_frame(mom_6d,1)
rel_mom_7d = standardize_data_frame(mom_7d,1)
rel_mom_15d = standardize_data_frame(mom_15d,1)
rel_mom_30d = standardize_data_frame(mom_30d,1)
rel_mom_50d = standardize_data_frame(mom_50d,1)
rel_mom_75d = standardize_data_frame(mom_75d,1)
rel_mom_100d = standardize_data_frame(mom_100d,1)
rel_mom_125d = standardize_data_frame(mom_125d,1)
rel_mom_150d = standardize_data_frame(mom_150d,1)
rel_mom_175d = standardize_data_frame(mom_175d,1)
rel_mom_200d = standardize_data_frame(mom_200d,1)
rel_mom_225d = standardize_data_frame(mom_225d,1)
rel_mom_250d = standardize_data_frame(mom_250d,1)

### TEST WHETHER MOMENTUM OR ABSOLUTE VALUE OF MOMENTUM IS MORE IMPORTANT
rel_abs_mom_1d = standardize_data_frame(abs_mom_1d,1)
rel_abs_mom_2d = standardize_data_frame(abs_mom_2d,1)
rel_abs_mom_3d = standardize_data_frame(abs_mom_3d,1)
rel_abs_mom_4d = standardize_data_frame(abs_mom_4d,1)
rel_abs_mom_5d = standardize_data_frame(abs_mom_5d,1)
rel_abs_mom_6d = standardize_data_frame(abs_mom_6d,1)
rel_abs_mom_7d = standardize_data_frame(abs_mom_7d,1)
rel_abs_mom_15d = standardize_data_frame(abs_mom_15d,1)
rel_abs_mom_30d = standardize_data_frame(abs_mom_30d,1)
rel_abs_mom_50d = standardize_data_frame(abs_mom_50d,1)
rel_abs_mom_75d = standardize_data_frame(abs_mom_75d,1)
rel_abs_mom_100d = standardize_data_frame(abs_mom_100d,1)
rel_abs_mom_125d = standardize_data_frame(abs_mom_125d,1)
rel_abs_mom_150d = standardize_data_frame(abs_mom_150d,1)
rel_abs_mom_175d = standardize_data_frame(abs_mom_175d,1)
abs_mom_200d = standardize_data_frame(abs_mom_200d,1)
abs_mom_225d = standardize_data_frame(abs_mom_225d,1)
abs_mom_250d = standardize_data_frame(abs_mom_250d,1)

counter = 0  
    
### NOW DO A GRID SEARCH OVER LAMBDA (PENALTY PARAMETER)
### FOR LAMBDA ON TRAINING SET PICK BETA
### FOR (LAMBDA_HAT, BETA_HAT) PICK MSE ON VALIDATION SET
### COMPUTE MSE OF THE ESTIMATOR ON THE TESTING SET TO SEE HOW PERFORMS    

for year in range(1999,2007):
    for month in range(1,13):
        print year, month
        start_train_date = pd.to_datetime(str(year) + '-' + str(month) + '-1') ### DAY TO START TRAINING
        if start_train_date <= pd.to_datetime('2006-08-01'):  ### LATEST TRAINING DATE FOR WHICH WE CAN GET TWO FULL YEARS IN
            end_train_date = start_train_date + timedelta64(12,'M') ### 12 MONTHS LATER
            start_valid_date = end_train_date + timedelta64(1,'D') ### COULD LENGTHEN THIS TO ELIMINATE DEPEDENCE
            end_valid_date = end_train_date + timedelta64(12,'M')
            ### NOW RUN ON A GRID SEARCH OVER THE PARAMETER SPACE
            ### THEN STORE THE OPTIMAL COEFFICIENTS, AS WELL AS THE MSE, R^2, AND MAE FOR EACH OF THE TESTING SETS  
            ## CONSTRUCT THE TRAINING AND VALIDATION MATRICES
            Xmat_tr, ymat_tr, Xmat_val, ymat_val = make_train_valid_matrices(start_train_date,end_train_date,start_valid_date,end_valid_date,fwd_vol_30d,mom_1d,mom_2d,mom_3d,mom_4d,mom_5d,mom_6d,mom_7d,mom_15d,mom_30d,mom_50d,mom_75d,mom_100d,mom_125d,mom_150d,mom_175d,mom_200d,mom_225d,mom_250d,abs_mom_1d,abs_mom_2d,abs_mom_3d,abs_mom_4d,abs_mom_5d,abs_mom_6d,abs_mom_7d,abs_mom_15d,abs_mom_30d,abs_mom_50d,abs_mom_75d,abs_mom_100d,abs_mom_125d,abs_mom_150d,abs_mom_175d,abs_mom_200d,abs_mom_225d,abs_mom_250d,back_vol_2d,back_vol_3d,back_vol_4d,back_vol_5d,back_vol_6d,back_vol_7d,back_vol_15d,back_vol_30d,back_vol_50d,back_vol_75d,back_vol_100d,back_vol_125d,back_vol_150d,back_vol_175d,back_vol_200d,back_vol_225d,back_vol_250d)
            Xmat_tr_withciv, ymat_tr_withciv, Xmat_val_withciv, ymat_val_withciv = make_train_valid_matrices_withcommonvars(start_train_date,end_train_date,start_valid_date,end_valid_date,fwd_vol_30d,civ_data_dates_aligned,mom_1d,mom_2d,mom_3d,mom_4d,mom_5d,mom_6d,mom_7d,mom_15d,mom_30d,mom_50d,mom_75d,mom_100d,mom_125d,mom_150d,mom_175d,mom_200d,mom_225d,mom_250d,abs_mom_1d,abs_mom_2d,abs_mom_3d,abs_mom_4d,abs_mom_5d,abs_mom_6d,abs_mom_7d,abs_mom_15d,abs_mom_30d,abs_mom_50d,abs_mom_75d,abs_mom_100d,abs_mom_125d,abs_mom_150d,abs_mom_175d,abs_mom_200d,abs_mom_225d,abs_mom_250d,back_vol_2d,back_vol_3d,back_vol_4d,back_vol_5d,back_vol_6d,back_vol_7d,back_vol_15d,back_vol_30d,back_vol_50d,back_vol_75d,back_vol_100d,back_vol_125d,back_vol_150d,back_vol_175d,back_vol_200d,back_vol_225d,back_vol_250d) 
            ### INITIALIZE THE GRID AND RUN THE REGRESSIONS
            ### RIDGE REGRESSION
            mse_ridge = np.zeros((len(ridge_grid),1))
            mae_ridge = np.zeros((len(ridge_grid),1))
            r2_ridge = np.zeros((len(ridge_grid),1))
            ridge_coeff_store = np.zeros((len(ridge_grid),n_predict))
            
            mse_ridge_withciv = np.zeros((len(ridge_grid),1))
            mae_ridge_withciv = np.zeros((len(ridge_grid),1))
            r2_ridge_withciv = np.zeros((len(ridge_grid),1))
            ridge_coeff_store_withciv = np.zeros((len(ridge_grid),n_predict+num_predict_civ))
            
            for ridge_index in range(len(ridge_grid)):
                clf_ridge = linear_model.Ridge (alpha=ridge_grid[ridge_index])
                ridgefit = clf_ridge.fit(Xmat_tr,ymat_tr)
                Ridge(alpha=ridge_grid[ridge_index], copy_X=True, fit_intercept=True, max_iter=None,normalize=False, solver='auto', tol=0.001)
                ridge_val = ridgefit.predict(Xmat_val)
                mse_ridge[ridge_index] = sklearn.metrics.mean_squared_error(ymat_val, ridge_val)
                mae_ridge[ridge_index] = sklearn.metrics.mean_absolute_error(ymat_val,ridge_val)
                r2_ridge[ridge_index] = sklearn.metrics.r2_score(ymat_val,ridge_val)
                ridge_coeff_store[ridge_index,:] = ridgefit.coef_

                clf_ridge_civ = linear_model.Ridge (alpha=ridge_grid[ridge_index])
                ridgefit_civ = clf_ridge_civ.fit(Xmat_tr_withciv,ymat_tr_withciv)
                Ridge(alpha=ridge_grid[ridge_index], copy_X=True, fit_intercept=True, max_iter=None,normalize=False, solver='auto', tol=0.001)
                ridge_val_civ = ridgefit_civ.predict(Xmat_val_withciv)
                mse_ridge_withciv[ridge_index] = sklearn.metrics.mean_squared_error(ymat_val_withciv, ridge_val_civ)
                mae_ridge_withciv[ridge_index] = sklearn.metrics.mean_absolute_error(ymat_val_withciv,ridge_val_civ)
                r2_ridge_withciv[ridge_index] = sklearn.metrics.r2_score(ymat_val_withciv,ridge_val_civ)
                ridge_coeff_store_withciv[ridge_index,:] = ridgefit_civ.coef_
                
            mse_lasso = np.zeros((len(lasso_grid),1))
            mae_lasso = np.zeros((len(lasso_grid),1))
            r2_lasso = np.zeros((len(lasso_grid),1))
            lasso_coeff_store = np.zeros((len(ridge_grid),n_predict))
            
            mse_lasso_withciv = np.zeros((len(lasso_grid),1))
            mae_lasso_withciv = np.zeros((len(lasso_grid),1))
            r2_lasso_withciv = np.zeros((len(lasso_grid),1))
            lasso_coeff_store_withciv = np.zeros((len(ridge_grid),n_predict+num_predict_civ))
            
            for lasso_index in range(len(lasso_grid)):
                clf_lasso = linear_model.Lasso (alpha=lasso_grid[lasso_index])
                lassofit = clf_lasso.fit(Xmat_tr,ymat_tr)
                Lasso(alpha=lasso_grid[lasso_index], copy_X=True, fit_intercept=True, max_iter=None,normalize=False,positive=False,precompute='Auto',tol=0.001,warm_start=False)
                lasso_val = lassofit.predict(Xmat_val)
                mse_lasso[lasso_index] = sklearn.metrics.mean_squared_error(ymat_val, lasso_val)
                mae_lasso[lasso_index] = sklearn.metrics.mean_absolute_error(ymat_val,lasso_val)
                r2_lasso[lasso_index] = sklearn.metrics.r2_score(ymat_val,lasso_val)
                lasso_coeff_store[lasso_index,:] = lassofit.coef_
                
                clf_lasso_civ = linear_model.Lasso (alpha=lasso_grid[lasso_index])
                lassofit_civ = clf_lasso_civ.fit(Xmat_tr_withciv,ymat_tr_withciv)
                Lasso(alpha=lasso_grid[lasso_index], copy_X=True, fit_intercept=True, max_iter=None,normalize=False,positive=False,precompute='Auto',tol=0.001,warm_start=False)
                lasso_val_civ = lassofit_civ.predict(Xmat_val_withciv)
                mse_lasso_withciv[lasso_index] = sklearn.metrics.mean_squared_error(ymat_val_withciv, lasso_val_civ)
                mae_lasso_withciv[lasso_index] = sklearn.metrics.mean_absolute_error(ymat_val_withciv,lasso_val_civ)
                r2_lasso_withciv[lasso_index] = sklearn.metrics.r2_score(ymat_val_withciv,lasso_val_civ)
                lasso_coeff_store_withciv[lasso_index,:] = lassofit_civ.coef_ 
                
            ### THINK ABOUT HOW TO START THE SEARCH AT THE CURRENT POINT - DISCUSSED IN CLASS TODAY - FOR EFFICIENCY.   
            ### STORE THE BEST LAMBDA_LASSO, LAMBDA_RIDGE    
            opt_mse_ridge_pos = argmin(mse_ridge)
            opt_mse_lasso_pos = argmin(mse_lasso)
            opt_mse_ridge_pos_withciv = argmin(mse_ridge_withciv)
            opt_mse_lasso_pos_withciv = argmin(mse_lasso_withciv)
            
            ### STORE THE OPTIMAL MSE, MAE, R-Squared
            train_lasso_rmse_store[counter,0] = mse_lasso[opt_mse_lasso_pos]**0.5
            train_ridge_rmse_store[counter,0] = mse_ridge[opt_mse_ridge_pos]**0.5
            train_lasso_r2_store[counter,0] = r2_lasso[opt_mse_lasso_pos]
            train_ridge_r2_store[counter,0] = r2_ridge[opt_mse_ridge_pos]
            train_lasso_mae_store[counter,0] = mae_lasso[opt_mse_lasso_pos]
            train_ridge_mae_store[counter,0] = mae_ridge[opt_mse_ridge_pos]
            
            train_lasso_rmse_store_withciv[counter,0] = mse_lasso_withciv[opt_mse_lasso_pos_withciv]**0.5
            train_ridge_rmse_store_withciv[counter,0] = mse_ridge_withciv[opt_mse_ridge_pos_withciv]**0.5
            train_lasso_r2_store_withciv[counter,0] = r2_lasso_withciv[opt_mse_lasso_pos_withciv]
            train_ridge_r2_store_withciv[counter,0] = r2_ridge_withciv[opt_mse_ridge_pos_withciv]
            train_lasso_mae_store_withciv[counter,0] = mae_lasso_withciv[opt_mse_lasso_pos_withciv]
            train_ridge_mae_store_withciv[counter,0] = mae_ridge_withciv[opt_mse_ridge_pos_withciv]
            
            lasso_lambda_store[counter,0] = lasso_grid[opt_mse_lasso_pos]
            ridge_lambda_store[counter,0] = ridge_grid[opt_mse_ridge_pos]
            lasso_opt_coeff_store[counter,:] = lasso_coeff_store[opt_mse_lasso_pos,:]
            ridge_opt_coeff_store[counter,:] = ridge_coeff_store[opt_mse_ridge_pos,:]     

            lasso_lambda_store_withciv[counter,0] = lasso_grid[opt_mse_lasso_pos_withciv]
            ridge_lambda_store_withciv[counter,0] = ridge_grid[opt_mse_ridge_pos_withciv]
            lasso_opt_coeff_store_withciv[counter,:] = lasso_coeff_store_withciv[opt_mse_lasso_pos_withciv,:]
            ridge_opt_coeff_store_withciv[counter,:] = ridge_coeff_store_withciv[opt_mse_ridge_pos_withciv,:]
            
            opt_ridge_lam = ridge_grid[opt_mse_ridge_pos]
            opt_ridge = linear_model.Ridge (alpha=opt_ridge_lam)
            ridgefit = opt_ridge.fit(Xmat_tr,ymat_tr) 
            Ridge(alpha=opt_ridge_lam, copy_X=True, fit_intercept=True, max_iter=None,normalize=False, solver='auto', tol=0.001)
            
            opt_lasso_lam = lasso_grid[opt_mse_lasso_pos]
            opt_lasso = linear_model.Lasso (alpha=opt_lasso_lam)
            lassofit = opt_lasso.fit(Xmat_tr,ymat_tr) 
            Lasso(alpha=opt_lasso_lam, copy_X=True, fit_intercept=True, max_iter=None,normalize=False,positive=False,precompute='Auto',tol=0.001,warm_start=False)          

            opt_ridge_lam_withciv = ridge_grid[opt_mse_ridge_pos_withciv]
            opt_ridge_withciv = linear_model.Ridge (alpha=opt_ridge_lam_withciv)
            ridgefit_withciv = opt_ridge_withciv.fit(Xmat_tr_withciv,ymat_tr_withciv) 
            Ridge(alpha=opt_ridge_lam_withciv, copy_X=True, fit_intercept=True, max_iter=None,normalize=False, solver='auto', tol=0.001)
            
            opt_lasso_lam_withciv = lasso_grid[opt_mse_lasso_pos_withciv]
            opt_lasso_withciv = linear_model.Lasso (alpha=opt_lasso_lam_withciv)
            lassofit_withciv = opt_lasso_withciv.fit(Xmat_tr_withciv,ymat_tr_withciv) 
            Lasso(alpha=opt_lasso_lam_withciv, copy_X=True, fit_intercept=True, max_iter=None,normalize=False,positive=False,precompute='Auto',tol=0.001,warm_start=False)          
            
            ### SEE HOW THE OPTIMAL CHOICE DID ON THE VARIOUS DATES
            y_test_set = [ytm_910, ytm_1011, ytm_1112, ytm_1213, ytm_1314]
            X_test_set = [Xtm_910, Xtm_1011, Xtm_1112, Xtm_1213, Xtm_1314]
            y_test_set_withciv = [ytm_910_withciv, ytm_1011_withciv, ytm_1112_withciv, ytm_1213_withciv, ytm_1314_withciv]
            X_test_set_withciv = [Xtm_910_withciv, Xtm_1011_withciv, Xtm_1112_withciv, Xtm_1213_withciv, Xtm_1314_withciv]            
            
            test_iter = 0
            for ytest,Xtest,ytest_civ,Xtest_civ in zip(y_test_set,X_test_set,y_test_set_withciv,X_test_set_withciv):
                
                lasso_predict = lassofit.predict(Xtest)
                lasso_rmse_store[counter,test_iter] = sklearn.metrics.mean_squared_error(ytest, lasso_predict)**0.5
                lasso_mae_store[counter,test_iter] = sklearn.metrics.mean_absolute_error(ytest,lasso_predict)
                lasso_r2_store[counter,test_iter] = sklearn.metrics.r2_score(ytest,lasso_predict)

                ridge_predict = ridgefit.predict(Xtest)
                ridge_rmse_store[counter,test_iter] = sklearn.metrics.mean_squared_error(ytest, ridge_predict)**0.5
                ridge_mae_store[counter,test_iter] = sklearn.metrics.mean_absolute_error(ytest,ridge_predict)
                ridge_r2_store[counter,test_iter] = sklearn.metrics.r2_score(ytest,ridge_predict)
                
                lasso_predict_withciv = lassofit_withciv.predict(Xtest_civ)
                lasso_rmse_store_withciv[counter,test_iter] = sklearn.metrics.mean_squared_error(ytest_civ, lasso_predict_withciv)**0.5
                lasso_mae_store_withciv[counter,test_iter] = sklearn.metrics.mean_absolute_error(ytest_civ,lasso_predict_withciv)
                lasso_r2_store_withciv[counter,test_iter] = sklearn.metrics.r2_score(ytest_civ,lasso_predict_withciv)

                ridge_predict_withciv = ridgefit_withciv.predict(Xtest_civ)
                ridge_rmse_store_withciv[counter,test_iter] = sklearn.metrics.mean_squared_error(ytest_civ, ridge_predict_withciv)**0.5
                ridge_mae_store_withciv[counter,test_iter] = sklearn.metrics.mean_absolute_error(ytest_civ,ridge_predict_withciv)
                ridge_r2_store_withciv[counter,test_iter] = sklearn.metrics.r2_score(ytest_civ,ridge_predict_withciv)
                
                naive_predict_7 = Xtest[:,42]
                naive_predict_15 = Xtest[:,43]
                naive_predict_30 = Xtest[:,44]
                naive_predict_50 = Xtest[:,45]
                naive_predict_75 = Xtest[:,46]
                
                naive_7_rmse_store[counter,test_iter] = sklearn.metrics.mean_squared_error(ytest,naive_predict_7)**0.5
                naive_7_mae_store[counter,test_iter] = sklearn.metrics.mean_squared_error(ytest,naive_predict_7)
                naive_7_r2_store[counter,test_iter] = sklearn.metrics.r2_score(ytest,naive_predict_7)

                naive_15_rmse_store[counter,test_iter] = sklearn.metrics.mean_squared_error(ytest,naive_predict_15)**0.5
                naive_15_mae_store[counter,test_iter] = sklearn.metrics.mean_squared_error(ytest,naive_predict_15)
                naive_15_r2_store[counter,test_iter] = sklearn.metrics.r2_score(ytest,naive_predict_15)

                naive_30_rmse_store[counter,test_iter] = sklearn.metrics.mean_squared_error(ytest,naive_predict_30)**0.5
                naive_30_mae_store[counter,test_iter] = sklearn.metrics.mean_squared_error(ytest,naive_predict_30)
                naive_30_r2_store[counter,test_iter] = sklearn.metrics.r2_score(ytest,naive_predict_30)

                naive_50_rmse_store[counter,test_iter] = sklearn.metrics.mean_squared_error(ytest,naive_predict_50)**0.5
                naive_50_mae_store[counter,test_iter] = sklearn.metrics.mean_squared_error(ytest,naive_predict_50)
                naive_50_r2_store[counter,test_iter] = sklearn.metrics.r2_score(ytest,naive_predict_50)

                naive_75_rmse_store[counter,test_iter] = sklearn.metrics.mean_squared_error(ytest,naive_predict_75)**0.5
                naive_75_mae_store[counter,test_iter] = sklearn.metrics.mean_squared_error(ytest,naive_predict_75)
                naive_75_r2_store[counter,test_iter] = sklearn.metrics.r2_score(ytest,naive_predict_75)
                
                test_iter+=1
            counter += 1
```
