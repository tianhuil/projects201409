```python
'''
Get the vix, fix futures data, and sp500 data we need from quandl, construct volatility and 
momentum measures over various horizons, save the final output
'''
import Quandl
import cPickle as pickle
import pandas as pd

start_date = '1999-01-01'
### GET THE VIX FUTURES DATA FOR USE IN CONSTRUCTING A TRADING STRATEGY
'''
INSERT QUANDL TOKEN HERE
'''
qdltoken = "" 

vix_data_fullsamp = Quandl.get(['YAHOO/INDEX_VIX.6'],authtoken=qdltoken)
sp500_data_fullsamp = Quandl.get(['YAHOO/INDEX_GSPC.6'],authtoken=qdltoken)
futures_vix_data = {}
fut_month = [1,2,3,4,5,6]
k=0
while k<len(fut_month):
    futures_vix_data[fut_month[k]] = Quandl.get(['OFDP/FUTURE_VX'+str(fut_month[k])+'.4'],authtoken=qdltoken)
    k+=1

'''
SOME OF THE DATA ARE MISSING IN CONTINUOUS CONTRACT SERIES, BUT THE DATA IS AVAILABLE
THE SPECIFIC CONTRACTS WHICH ARE ALSO AVAILABLE.  NOTE THAT IF DO TRADING FOR LONGER THAN
ONE MONTH HORIZON, WILL NEED TO INSERT THESE VALUES FOR THE APPROPRIATE TIME-FRAME
'''
futures_vix_data[1].loc[pd.to_datetime('2013-05-24')] = 15.5
futures_vix_data[1].loc[pd.to_datetime('2013-06-04')] = 16.1
futures_vix_data[1].loc[pd.to_datetime('2013-06-07')] = 15.8
futures_vix_data[1].loc[pd.to_datetime('2013-06-28')] = 18.05
futures_vix_data[1].loc[pd.to_datetime('2013-09-20')] = 15.0
futures_vix_data[1].loc[pd.to_datetime('2013-09-23')] = 15.35
futures_vix_data[1].loc[pd.to_datetime('2013-10-10')] = 16.4
futures_vix_data[1].loc[pd.to_datetime('2013-10-11')] = 15.7
futures_vix_data[1].loc[pd.to_datetime('2014-02-14')] = 14.2

vix_data = vix_data_fullsamp.loc[vix_data_fullsamp.index>=start_date]
sp500_data = sp500_data_fullsamp.loc[sp500_data_fullsamp.index>=start_date]

'''Compute the various EWMA for use in constructing momentum'''
ewma_sp_30d = pd.stats.moments.ewma(sp500_data, com=14.5) 
ewma_sp_15d = pd.stats.moments.ewma(sp500_data, com=7)  
ewma_sp_5d = pd.stats.moments.ewma(sp500_data, com=2.0)  

def get_mom(smoothed_price,lag):
    '''GET THE DIFFERENCE IN THE SMOOTHED MOVING AVERAGE BETWEEN PREVIOUS TRADING DAY AND lag PERIODS AGO, 
    OR THE FEWEST NUMBER OF DAYS STRICTLY GREATER THAN lag FOR WHICH PRICE DATA IS AVAILABLE'''
    mom = pd.DataFrame(data=None,index=smoothed_price.index,columns=smoothed_price.columns)
    for dates in smoothed_price.index:
        counter=0
        while True:
            if min(smoothed_price.index)>(dates - timedelta64(lag+counter,'D')):               
                break
            close_ndays = lag+counter
            date_match = dates-timedelta64(close_ndays,'D')
            if any(smoothed_price.index==date_match):
                mom.loc[dates]= 100.0*365.0/close_ndays*(smoothed_price.loc[dates]-smoothed_price.loc[date_match])/smoothed_price.loc[date_match]               
                break
            else:
                counter+=1           
    return mom

def get_fwd_vol(price,num_days):
    ### GET VOLATILITY COMPUTED FROM THE CURRENT DATE TO CURRENT DATE + num_days IN THE FUTURE
    returns = price.pct_change()
    fwd_vol = pd.DataFrame(data=None,index=price.index,columns=price.columns)
    for dates in price.index:
        fwd_flag=dates+timedelta64(num_days,'D')
        good_index = logical_and((returns.index<=fwd_flag), (returns.index>=dates))
        fwd_vol.loc[dates] = std(100.0*(252.0**0.5)*returns.loc[returns.index[good_index]])
    return fwd_vol

def get_back_vol(price,num_days):
    ### GET VOLATILITY COMPUTED FROM THE CURRENT DATE TO CURRENT DATE - 1 TO CURRENT DATE-1-num_days IN THE PAST
    returns = price.pct_change()
    back_vol = pd.DataFrame(data=None,index=price.index,columns=price.columns)
    for dates in price.index:
        back_flag=dates-timedelta64(num_days,'D')
        good_index = logical_and((returns.index>=back_flag), (returns.index<dates))
        back_vol.loc[dates] = std(100.0*(252.0**0.5)*returns.loc[returns.index[good_index]])
    return back_vol

sp_fwd_vol = {}
num_days_fwd_vol = [30,60,90,120,150,180,210,250]
k=0
while k<len(num_days_fwd_vol):
    sp_fwd_vol[num_days_fwd_vol[k]] = get_fwd_vol(sp500_data,num_days_fwd_vol[k])
    k+=1
    
sp_back_vol = {}
sp_mom_5dewma = {}
sp_mom_15dewma = {}
sp_mom_30dewma = {}
sp_abs_mom_5dewma = {}
sp_abs_mom_15dewma = {}
sp_abs_mom_30dewma = {}

num_days_back_vol = [2,3,4,5,6,7,15,30,50,75,100,125,175,200,225,250]  
num_days_mom = [1,2,3,4,5,6,7,15,30,50,75,100,125,175,200,225,250]
k=0
while k<len(num_days_back_vol):
    sp_back_vol[num_days_back_vol[k]] = get_back_vol(sp500_data,num_days_back_vol[k])
    k+=1
    
k=0   
while k<len(num_days_mom):
    sp_mom_5dewma[num_days_mom[k]] = get_mom(ewma_sp_5d,num_days_mom[k])
    sp_mom_15dewma[num_days_mom[k]] = get_mom(ewma_sp_15d,num_days_mom[k])
    sp_mom_30dewma[num_days_mom[k]] = get_mom(ewma_sp_30d,num_days_mom[k])
    
    sp_abs_mom_5dewma[num_days_mom[k]] = abs(sp_mom_5dewma[num_days_mom[k]])
    sp_abs_mom_15dewma[num_days_mom[k]] = abs(sp_mom_15dewma[num_days_mom[k]])
    sp_abs_mom_30dewma[num_days_mom[k]] = abs(sp_mom_30dewma[num_days_mom[k]])
    k+=1

data_list = [futures_vix_data,sp_fwd_vol, sp_back_vol,sp_mom_5dewma,sp_mom_15dewma,sp_mom_30dewma, \
             sp_abs_mom_5dewma,sp_abs_mom_15dewma,sp_abs_mom_30dewma,vix_data,sp500_data]
data_list_save = ["futures_vix_data_save.dat","sp_fwd_vol_save.dat", "sp_back_vol_save.dat","sp_mom_5dewma_save.dat","sp_mom_15dewma_save.dat","sp_mom_30dewma_save.dat", \
             "sp_abs_mom_5dewma_save.dat","sp_abs_mom_15dewma_save.dat","sp_abs_mom_30dewma_save.dat","vix_data_save.dat","sp500_data_save.dat"]

for items,save_items in zip(data_list,data_list_save):
    with open(save_items, "wb") as f:
        pickle.dump(items, f)
 
```
