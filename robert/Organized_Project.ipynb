```python
num_rows=np.max(df_hashed['restaurant_id'])
num_cols=np.max(df_hashed["user_id"])

#forming restaurant x user matrix of ratings

def matrix_generator(df1 , rating):
        df=df1

        elements = df[rating]
        rows = df['restaurant_id']
        cols = df['user_id']

        resto_user_array=SP.coo_matrix((elements,(rows,cols)),shape=(num_rows+1,num_cols+1))
        
        return resto_user_array

full_resto_user_array = matrix_generator(df_hashed, 'rating')
full_normed_resto_user_array = matrix_generator(df_hashed, 'normed_rating')


#def reverse_transformer(array):
#    array
```

```python
#the nearest neighbors estimator:

from sklearn.base import BaseEstimator, ClassifierMixin

class NearestNeighborsEstimator(BaseEstimator, ClassifierMixin):
    
    def __init__(self, shrink_coeff):
        self.shrink_coeff=shrink_coeff
        pass
    
    def fit(self, array,normed_array):
        #takes as input resto x user arrays
        normed_array.tocsr()
        array.tocsr()
        sqrsum = (normed_array.multiply(normed_array)).sum(axis=1)
        sqrdnm = np.outer(sqrsum,sqrsum)
        dnm = np.sqrt(sqrdnm)
        self.corr_matrix=(normed_array.dot(normed_array.T)).toarray()/dnm
        
        ones=SP.coo_matrix((np.ones(array.nnz),array.nonzero()))
        ones=ones.tocsr()
        counts= ones.dot(ones.T)
        counts = counts.todense()
        
        self.shrink = (counts-1)/(counts - 1 + self.shrink_coeff)
        
        self.sim_matrix = self.corr_matrix * self.shrink.getA()
        
        return self
      


    def predict(self, test_array):
        simsparse= SP.csr_matrix(self.sim_matrix)
        sim_numerator = simsparse.dot(test_array) 

        test_array_nonzero_ones = SP.coo_matrix((np.ones(test_array.nnz),test_array.nonzero())).tocsr()
        sim_support_sum=simsparse.dot(test_array_nonzero_ones) - test_array #sum of all sim measures rated by the user not counting current one
        diag=SP.diags(simsparse.diagonal(),0)
        sub = diag.dot(test_array)

        self.predictions_nobaseline = SP.csr_matrix((sim_numerator / sim_support_sum) - sub)
        
        #for scoring purposes, only want to test against ratings actually in the dataset. Thus, we return the above predictions
        #only for the ratings actually in the data:
        
        self.score_predictions=self.predictions_nobaseline.multiply(test_array_nonzero_ones)
        
        return self.score_predictions
    
        #can ask for predictions_nobaseline to get predictions, but it only returns these for score
        
        #Need to add baseline prediction to get final prediction. But simply for cross-validation etc, dont need to worry about this
        #Can just compute scores on residuals and then for final prediction (as in output of pipeline) would add them together        
        
        
        
        
        
```

# Other Analysis

None of the below anaylsis made it into the final version of the project, but is still quite interesting.

My goals after completing the nearest neighbor's analysis was two-fold. First, I wanted to see if I could improve my recommendations using alternate techniques. Second, I wanted to see if I could utilize the actual review text that I had scraped in order to tease out subtle information about HOW the recommended restaurants where related. 

For the first goal, I performed a latent factor analysis on the residuals of the baseline predictors. This entailed taking the restaurant x user matrix of ratings, and performing SVD on it with a specified number of factors. Unfortunately, this analysis did not perform as well as the nearest neighbors analysis. For any reasonable number of factors, the predictive capababilities of the factorized model were significantly worse than the neighbor's based predictions. Futher, they lacked the explainability of the nearest neighbors model.

I think the latent factorization technique failed because the ratings are hyper-local in the sense that users tend to primarily rate restaurants from a very small geographic area. As a result, we ended up identifying factors that basically described the locations of restaurants in a very coarse manner, but did not add at all to the prediction.


FOr the second goal, in an attempt to "identify" the topics of underlying factors determining restaurant similarity, I adopted the following strategy:

I performed PCA on the matrix of restaurant similarity scores output by the nearest neighbors model. The idea was that this would suffer less from the geographic issues associated with factorizing the ratings with only the baselines removed.
I then look at each individual factor, ordered restaurants according to this factor and then formed bins of review text based on this ordering.

As an example, consider the first factor (i.e. the most impactful one) I then ordered the restuarants according to their score in this dimension and binned this ordered list of restaurants (I varied the sizes of bins from 20 to 500 restaurants). 
I then considered all the review text associated with the restaurants in a bin to be a single document. I only included review texts with 4 and 5 star ratings. I then performed TF-IDF, both comparing the top bins across all factors and comparing the top bin against the bottom bin for any single factor. 

I also performed the same analysis, using latent dirichlet allocation to discover these topics as well.

The hope is that either of these methods can identify the words and n-grams asssociated with the latent factors determining the similarity sores between restuarants. (In some sense identifying these latent factors). My hope was to then be able to display the words and n-grams associated with the most important factors for each restaurant recommended by my engine as a way of giving the user more information about WHY exactly the two restaurants are similar.

Unfortunately, under the time constraints of the project, I have not yet figured out what is going wrong with the analysis.

```python
#Fit method will take a normed resto x user array as input. 
import scipy.sparse.linalg as LA
import scipy.linalg
#NOTE: This assumes baseline predictors already removed!!!!!


class latent_factor_model(BaseEstimator, ClassifierMixin):
    
    def __init__(self, n_factors):
        self.n_factors=n_factors
        pass
    
    def fit(self, normed_resto_user_train):
        (self.R,self.S,self.U) = LA.svds(normed_resto_user_train, self.n_factors)
        self.resto_factors = self.R.dot(scipy.linalg.sqrtm(np.diag(self.S)))
        self.user_factors = scipy.linalg.sqrtm(np.diag(self.S)).dot(self.U)
        return self
    
    #def predict(self, normed_test_df):
        
        
    
```
