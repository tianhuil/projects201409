# Analyzing and Predicting Crime in Chicago

In this project, I consider the Redeye Chicago homicide dataset that tracks homicides in Chicago sine 2007, and the crime dataset from the Chicago data portal, which contains over 5.5 million records of crime that occurred in Chicago since 2001.  Although the Chicago crime dataset also contains homicide data, the Redeye Chicago dataset contains much more detailed information such as the age and race of the victims. Along with the data on crimes, datasets from the census and weather provide information on demographics, socioeconomic factors and temperature.

The main motivations of this project are to provide an analysis of crime trends and patterns, and to provide a prediction of crime occurrence by location and time. The analysis reveals that overall crime is steadily decreasing, and drug and violent crimes have decreased the most in the past 10 years. The crimes are negatively correlated with income, that is, overall crime rate is higher is poor communities. However, the proportion of property crimes is higher in communities with high income.

The limited resources of the police department could be deployed much more effectively we could predict ahead of time when and where crimes are likely to occur. My crime prediction shows that past history of crime trends, in other words, when and where crimes tended to occurr in the past, joined with the daily temperature data, can provide quite well, with a Spearman's rank correlation coefficient of 0.8, guidance in how to deploy police officers across communities and shifts.

## Overall Trends in Crime

### Overall rate of crime in Chicago between 2004 and 2013.

```python
Image(filename = '../figs/crimes.png')
```

The overall crime rate in Chicago has been steady decreasing between the years 2004 and 2013. The proportion of property, violent, and drug crimes are broadly constant over the years, although the proportion of property crimes has increased slightly since 2008 while the proportion of drug crimes has decreased.

Violent crimes are are those committed directly against a person: Homicide, Criminal Sexual Assault, Robbery, and Aggravated Assault/Battery. Property crimes are those in which there is no direct threat or harm to a person - Burglary, Theft, Motor Vehicle Theft, Criminal Damage, and Arson.

```python
Image(filename = '../figs/crimes_change.png')
```

Most crimes have decresed in the 10 years to 2013, but fraud shows a small increase. Among crime types, drug crimes have decreased the most, while property crimes have decreased the least.

Most crime types show a broad decline over the years. However, different crimes show different patterns of occurrence. ### Which crimes tend to occur in certain communities?

```python
Image(filename = '../figs/crimes_area_grouped.png')
```

The plots indicate that the proportion of property crimes tends to increase with income, while the proportions of violent and drug crimes decrease with increasing income. This is corroborated by considering the spearman's rank correlation coefficient  between per capita income and proportion of a particular type of crime. The coefficient measures how well the relationship between two variables can be described using a monotonic function. A value of 1 or -1 indicates each of the variables is a perfect monotonically increasing or decreasing function of the other. The coefficients are 0.66, -0.65, and -0.71, respectively, for the correlation between per capita income vs. proportion of property, violent, and drug crimes.

### Crime Prediction

```python
Image('../figs/crimes_trend.png')
```

Dividing each day into three 8-hour shifts that closely follow actual police shifts, I consider the variation of crime rate on day of week and time of day. The week divides into 21 shifts, and the above plot shows that there is a clear pattern of crime rate being higher during the evening hours, and on weekends. Crimes rates are highest on Friday evenings. During the night (between midnight and 8 am), crime rates are highest on Saturdays and Sundays; however, crimes rates are lowest on those days between 8 am and midnight.

```python
Image('../figs/predict.png')
```

The prediction was based on a log regression of the following features: the community area, day of week, time of day (each day is divided into three shifts as above), the maximum and minimum temperatures of the day. The data from the previous one year period (8/4/2013 - 8/3/2014) was used to predict crime occurrence during this particular week. A perfect prediction would have the points lined up along the diagonal line of the plot.

```python
Image('../figs/summary.png')
```

The utility of predicting crimes lies in being able to deploy police officers ahead of time so as to maximize their effectiveness. A measure of how well the prediction performs in this sense is shown in the above plots. The figure on the left shows how well the prediction does in allocating police officers to different communities for each 8 hour shift during the week. Let's assume that the strategy is to select 10 out of the 77 communities of Chicago and concentrate all the police efforts upon these 10. For a particular 8 hour shift during the week of Aug. 4th, randomly selecting 10 communities would result in targeting only 12% of the total crimes that were committed during that shift, on average. Selecting the 10 communities with the highest predicted number of crimes results in 34% of the crimes being targeted, while the best-case scenario, when you know ahead of time which 10 communities would have the highest number of crimes committed, would result in 40% of the crimes being targeted, on average.

It is likely that the police officers in one community cannot easily be deployed in another. In this case, the figure on the right shows how best to allocate officers in each community to different shifts. For this scenario, let's assume that the strategy is to concentrate police efforts on 10 out of the 21 shifts during the week. For a particular community area, randomly selecting 10 shifts would target 50% of the total crimes committed during the week in that community area, on average. Selecting the 10 shifts with the highest predicted number of crimes results in 60% of the crimes being targeted, while the best-case scenario, where you know in which 10 shifts the highest number of crimes would be commited ahead of time, would result in 75% of crimes being targeted, on average.
