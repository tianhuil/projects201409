### Murder rates by race, gender, and age of victim

```python
Image(filename = '../figs/murders_grouped.png')
```

The race categories are Asian, African American, and White (White includes the hispanic/latino population). Gunshot is the main cause of homicides, with stabbing coming second. An African American is more than 7 times likely to be murdered in Chicago as a white person. Most of the African American murder victims are male; almost 80 out of every 100,000 African American men are murdered in Chicago each year, while the rate for white men is around 10 out of every 100,000. The rate for African American women is a tenth of that for African American men (about 8 per 100,000) and that for white women is 2 per every 100,000. For both men and women, the proportion of murders from gunshot wounds is higher for African Americans than for whites. African American men 15 and 24 of age is the most vulnerable group. Over 100 out of every 100,000 African American men in this age group gets murdered each year, 98% from gunshot wounds. Together, African American men of ages beteen 15 and 34 make up 56% of the total murder victims in Chicago.

### Murders by neighborhood

```python
Image(filename = '../figs/murders_ca.png')
```

Murders are strongly segregated, with just two community areas, Austin and South Shore, accounting for 15% of total murders that occurred in Chicago in 2013. ## Overall Trends in Crime

### Overall rate of crime in Chicago between 2004 and 2013.

```python
Image(filename = '../figs/crimes.png')
```

The overall crime rate in Chicago shows a steady decrease between the years 2004 and 2013. The proportion of property, violent, and drug crimes are broadly constant over the years, although the proportion of property crimes has increased slightly since 2008 while the proportion of drug crimes has decreased.

Violent crimes are are those committed directly against a person: Homicide, Criminal Sexual Assault, Robbery, and Aggravated Assault/Battery. Property crimes are those in which there is no direct threat or harm to a person - Burglary, Theft, Motor Vehicle Theft, Criminal Damage, and Arson.

### Which crimes have decreased the most/least between 2004 and 2013?

```python
Image(filename = '../figs/crimes_change.png')
```

Most crimes have decresed in the 10 years to 2013, but fraud shows a small increase. Among crime types, drug crimes have decreased the most, and property crimes have decreased the least.

`*`The population of Chicago for each year was obtained by linearly interpolating between the populations at 2000 and 2010.

Most crime types show a broad decline over the years. However, different crimes show different patterns of occurrence. For example, drug crime shows a marked increase in the years between 2003 and 2008. Assault/battery shows sharp peak in the years 2004 and 2006, and decreases rapidly in the following years. Burglary shows a broad increase between 2007 and 2011.
