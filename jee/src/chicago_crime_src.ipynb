## Where did the data come from?

```python
import pandas as pd

crimes = pd.read_csv('../data/crimes_2001_to_present.csv', usecols = [1, 2, 5, 8, 13, 19, 20])
```

```python
# Parse the date field
crimes['DateTime'] = crimes['Date'].map(pd.to_datetime)
crimes = crimes.drop('Date', axis = 1)

# Only consider data from 2004 (the last 10 years)
crimes = crimes[crimes['DateTime'] >= dt.datetime(2004, 1, 1)]

# Clean up community area field
crimes['Community Area'] = crimes['Community Area'].map(lambda x: 
                                                        int(x) if not (pd.isnull(x) or x == ' ') else 0)
crimes = crimes[crimes['Community Area'] != 0]
```

The crimes data was downloaded from the Chicago data portal, covering incidences of crime from 2001, and contains over 5.5 million records.

```python
import pandas as pd

# The demographics table was obtained by analyzing the 2010 census data for the city of Chicago.
# census data for age and sex
P12 = pd.read_csv('../data/DEC_10_SF1_P12_with_ann.csv', header = 1)
P12_Chicago = P12[P12['Geography'] == 'Chicago city, Illinois']

# census data for age and sex for non-hispanic whites
P12I = pd.read_csv('../data/DEC_10_SF1_P12I_with_ann.csv', header = 1)
P12I_Chicago = P12I[P12I['Geography'] == 'Chicago city, Illinois']

# census data for age and sex for hispanics
P12H = pd.read_csv('../data/DEC_10_SF1_P12H_with_ann.csv', header = 1)
P12H_Chicago = P12H[P12H['Geography'] == 'Chicago city, Illinois']

# census data for age and sex for blacks
P12B = pd.read_csv('../data/DEC_10_SF1_P12B_with_ann.csv', header = 1)
P12B_Chicago = P12B[P12B['Geography'] == 'Chicago city, Illinois']

# census data for age and sex for asians
P12D = pd.read_csv('../data/DEC_10_SF1_P12D_with_ann.csv', header = 1)
P12D_Chicago = P12D[P12D['Geography'] == 'Chicago city, Illinois']

group_names = ['Under 5', '5 to 14', '15 to 24', '25 to 34', '35 to 44', '45 to 54',
               '55 to 64', 'Over 65', 'Total']

# Regroup the census data into the above age groups
def group_ages(series):
    age_groups = []
    age_groups.append(series[1])
    age_groups.append(series[2:4].sum())
    age_groups.append(series[4:9].sum())
    age_groups.append(series[9:11].sum())
    age_groups.append(series[11:13].sum())
    age_groups.append(series[13:15].sum())
    age_groups.append(series[15:18].sum())
    age_groups.append(series[18:].sum())
    age_groups.append(series[0])
    return age_groups

P12_R = pd.DataFrame({'Male': group_ages(P12_Chicago.ix[:, 4:28].stack()), 
                      'Female': group_ages(P12_Chicago.ix[:, 28:].stack())}, index = group_names)
P12IH_R = pd.DataFrame({'Male': [sum(x) for x in zip(group_ages(P12I_Chicago.ix[:, 4:28].stack()), 
                                                     group_ages(P12H_Chicago.ix[:, 4:28].stack()))], 
                        'Female': [sum(x) for x in zip(group_ages(P12I_Chicago.ix[:, 28:].stack()), 
                                                       group_ages(P12H_Chicago.ix[:, 28:].stack()))]}, 
                       index = group_names)
P12B_R = pd.DataFrame({'Male': group_ages(P12B_Chicago.ix[:, 4:28].stack()), 
                       'Female': group_ages(P12B_Chicago.ix[:, 28:].stack())}, index = group_names)
P12D_R = pd.DataFrame({'Male': group_ages(P12D_Chicago.ix[:, 4:28].stack()), 
                       'Female': group_ages(P12D_Chicago.ix[:, 28:].stack())}, index = group_names)

demographics = pd.concat([P12_R, P12IH_R, P12B_R, P12D_R], axis = 0, 
                         keys = ['ALL', 'WHITE', 'AFAM', 'ASIAN'])

# The demographics by community area data was downloaded from http://robparal.blogspot.com/2012/05/hard-to-find-census-data-on-chicago.html.
names = ['Community Area', 'Neighborhood', 'NHW', 'NHB', 'NHAM', 'NHAS', 
         'NHOTHER', 'HISP', 'Multiple Race', 'Total']

census2000 = pd.io.excel.read_excel('../data/Race_by_Community_Area_2000_2010.xlsx', skiprows = [0, 1, 2, 3, 4], 
                                    parse_cols = 'A:J', names = names, header = None, skip_footer = 10,
                                    index_col = 'Community Area')
census2010 = pd.io.excel.read_excel('../data/Race_by_Community_Area_2000_2010.xlsx',skiprows = [0, 1, 2, 3, 4], 
                                    parse_cols = 'L:U', names = names, header = None, skip_footer = 10,
                                    index_col = 'Community Area')

comm_area = pd.read_csv('../data/CommAreas.csv', usecols = [4, 5, 7], index_col = 'AREA_NUMBE').sort_index()

# sum NHW and HISP columns into WHITE
census2000['White'] = census2000['NHW'] + census2000['HISP']
census2010['White'] = census2010['NHW'] + census2010['HISP']

# rename NHB to Black, and NHAS to Asian
census2000.rename(columns = {'NHB': 'AFAM', 'NHAS': 'ASIAN'}, inplace = True)
census2010.rename(columns = {'NHB': 'AFAM', 'NHAS': 'ASIAN'}, inplace = True)

# include population density field
# convert sq ft into sq mi
census2000['Population Density'] = census2000['Total'].astype(float)/comm_area['SHAPE_AREA']*27878400
census2010['Population Density'] = census2010['Total'].astype(float)/comm_area['SHAPE_AREA']*27878400

# drop unnecessary columns and rows
dropcols = ['NHW', 'NHAM', 'NHOTHER', 'HISP', 'Multiple Race']
census2000.drop(dropcols, axis = 1, inplace = True)
census2010.drop(dropcols, axis = 1, inplace = True)
census2000.drop('TOTAL', axis = 0, inplace = True)
census2010.drop('TOTAL', axis = 0, inplace = True)
```

The demographics table was obtained by analyzing the 2010 census data for the city of Chicago. It tabulates the population grouped into age brackets and sex. The tables referenced are DEC_10_SF1_P12_with_ann.csv, DEC_10_SF1_P12B_with_ann.csv, DEC_10_SF1_P12D_with_ann.csv, DEC_10_SF1_P12H_with_ann.csv, and DEC_10_SF1_P12I_with_ann.csv.

The census data that gives the demographics of Chicago community areas was downloaded from http://robparal.blogspot.com/2012/05/hard-to-find-census-data-on-chicago.html.

The total land area of community areas was downloaded from the chicago data portal.

```python
import pandas as pd
import dateutil.parser

def date_parser(date_str, time_str):
    if pd.isnull(time_str):
        time_str = ' '
    return dateutil.parser.parse(' '.join([date_str, time_str]))

names = ['Address', 'Date', 'Time', 'Location', 'Neighborhood', 'Age', 'Gender', 
         'Race', 'Name', 'Cause', 'Story URL', 'Case Number', 'Charges and Trials']
homicides2007 = pd.read_csv('../data/homicides_2007.csv', parse_dates = [['Date', 'Time']], 
                            date_parser = date_parser, skiprows = [0], names = names)
homicides2008 = pd.read_csv('../data/homicides_2008.csv', parse_dates = [['Date', 'Time']],
                            date_parser = date_parser, skiprows = [0], names = names)
homicides2009 = pd.read_csv('../data/homicides_2009.csv', parse_dates = [['Date', 'Time']],
                            date_parser = date_parser, skiprows = [0], names = names)
homicides2010 = pd.read_csv('../data/homicides_2010.csv', parse_dates = [['Date', 'Time']],
                            date_parser = date_parser, skiprows = [0], names = names)
homicides2011 = pd.read_csv('../data/homicides_2011.csv', parse_dates = [['Date', 'Time']],
                            date_parser = date_parser, skiprows = [0], names = names)
homicides2012 = pd.read_csv('../data/homicides_2012.csv', parse_dates = [['Date', 'Time']],
                            date_parser = date_parser, skiprows = [0], names = names)
homicides2013 = pd.read_csv('../data/homicides_2013.csv', parse_dates = [['Date', 'Time']],
                            date_parser = date_parser, skiprows = [0], names = names)

homicides = pd.concat([homicides2007, homicides2008, homicides2009, homicides2010,
                       homicides2011, homicides2012, homicides2013], axis = 0)
homicides['Year'] = homicides['Date_Time'].map(lambda x: x.year)

# process 'Age' field into a numeric 'Age_R' field.
def get_age(age_str):
    if pd.isnull(age_str): return None
    
    if age_str == 'Newborn':
        return 0
    elif ('day' in age_str) or ('week' in age_str) or ('month' in age_str):
        return 0
    elif '20s' in age_str:
        return 25
    elif '30s' in age_str:
        return 35
    else:
        return int(age_str)
    
homicides['Age_R'] = homicides['Age'].map(get_age)

age_groups = ['Under 5', '5 to 14', '15 to 24', '25 to 34', '35 to 44', '45 to 54',
              '55 to 64', 'Over 65']
ages = [0, 5, 15, 25, 35, 45, 55, 65, 100]

homicides['Age Group'] = pd.cut(homicides['Age_R'], ages, right = False, labels = age_groups)
homicides= homicides.drop('Age_R', axis = 1)
```

The homicides data were downloaded from the redeye Chicago homicide tracker website, http://homicides.redeyechicago.com/. The dataset includes 100 more homicides over the period 2007-2013 than the dataset from the Chicago data portal, and includes demographic information on the victims.

`*`In aggregating the murders into victim's age groups, there are 64 datapoints with missing ages. Those indicated 'unknown: 20s' (3 datapoints) and 'unknown: 30s' (1 datapoint) were reclassified as being of age 25 and 35, respectively.

### Murder rates by race, gender and age

```python
# The number of murders by race of victim
homicides['Race'] = homicides['Race'].str.rstrip()
homicides.replace('Black', 'AFAM', inplace = True)
homicides.replace('White', 'WHITE', inplace = True)
homicides.replace('Asian', 'ASIAN', inplace = True)
race = homicides.groupby(['Race']).size()

#gunshots = homicides[homicides['Cause'] == 'Gunshot'].groupby('Race').size()

pop = pd.Series([demographics.ix[('ASIAN', 'Total'), ].sum(),
                 demographics.ix[('AFAM', 'Total'), ].sum(),
                 demographics.ix[('WHITE', 'Total'), ].sum()], index = ['ASIAN', 'AFAM', 'WHITE'])

nyears = 7
#homicide_by_race = pd.concat([gunshots/nyears/pop*100000, race/nyears/pop*100000], 
#                             axis = 1, keys = ['Gunshot', 'Total'])

homicide_by_race = race/nyears/pop*100000
homicide_by_race.index.name = None

# The number of murders by race and gender of victim
gender = homicides.groupby(['Race', 'Gender']).size()

#gunshots_gender = homicides[homicides['Cause'] == 'Gunshot'].groupby(['Race', 'Gender']).size()

pop_gender = pd.concat([demographics.ix[('ASIAN', 'Total'), ], 
                        demographics.ix[('AFAM', 'Total'), ],
                        demographics.ix[('WHITE', 'Total'), ]], keys = ['ASIAN', 'AFAM', 'WHITE'])
pop_gender.index.names = ['Race', 'Gender']

nyears = 7
homicide_by_gender = (gender/nyears/pop_gender*100000).unstack('Gender')
homicide_by_gender.columns.name = None
homicide_by_gender.index.name = None

# The number of murders by race and age of victim
age = homicides.groupby(['Race', 'Age Group']).size()

#gunshots_age = homicides[homicides['Cause'] == 'Gunshot'].groupby(['Race', 'Age Group']).size()

pop_age = pd.concat([demographics.ix[('ASIAN', ), ].sum(1)[:-1],
                     demographics.ix[('AFAM', ), ].sum(1)[:-1],
                     demographics.ix[('WHITE', ), ].sum(1)[:-1]], keys = ['ASIAN', 'AFAM', 'WHITE'])
pop_age.index.names = ['Race', 'Age Group']

nyears = 7
#homicide_by_age = pd.concat([gunshots_age/nyears/pop_age*100000, 
#                             age/nyears/pop_age*100000], axis = 1, keys = ['Gunshot', 'Total']).fillna(0)

homicide_by_age = (age/nyears/pop_age*100000).fillna(0).unstack()
homicide_by_age.columns.name = None
cols = ['Under 5', '5 to 14', '15 to 24', '25 to 34', '35 to 44', '45 to 54', '55 to 64', 'Over 65']
homicide_by_age = homicide_by_age[cols]
homicide_by_age.index.name = None

# Coverage of murders in the media
media = (pd.isnull(homicides['Story URL'])) | (pd.isnull(homicides['Charges and Trials']))
coverage = homicides[~media].groupby(['Race']).size()/race
```

## Overall Trends in Crime

Violent index crimes are are those committed directly against a person: Homicide, Criminal Sexual Assault, Robbery, and Aggravated Assault/Battery.
Property index crimes are those in which there is no direct threat or harm to a person - Burglary, Theft, Motor Vehicle Theft, Criminal Damage, and Arson.

In Part II, the following categories are tracked: simple assault, curfew offenses and loitering, embezzlement, forgery and counterfeiting, disorderly conduct, driving under the influence, drug offenses, fraud, gambling, liquor offenses, offenses against the family, prostitution, public drunkenness, runaways, sex offenses, stolen property, vandalism, vagrancy, and weapons offenses.

```python
crimes['Year'] = crimes['DateTime'].map(lambda x: x.year)
```

```python
import pandas as pd

crimes_data = crimes.copy()
crimes_data = crimes_data[crimes_data['Year'] < 2014]

# Divide the crimes into categories of violent, property, drug, and other.
Violent_Crime = ['HOMICIDE', 'CRIM SEXUAL ASSAULT', 'ASSAULT', 'BATTERY', 'ROBBERY']
Property_Crime = ['THEFT', 'BURGLARY', 'CRIMINAL DAMAGE', 'MOTOR VEHICLE THEFT', 'ARSON']
Drug_Crime = ['NARCOTICS']

def get_category(type):
    if type in Violent_Crime:
        return 'Violent'
    elif type in Property_Crime:
        return 'Property'
    elif type in Drug_Crime:
        return 'Drug'
    else:
        return 'Other'
    
crimes_data['Category'] = crimes_data['Primary Type'].apply(get_category)

# The population of each year is interpolated between the populations of 2000 and 2010
from scipy.interpolate import InterpolatedUnivariateSpline 

yr = np.linspace(2001, 2010, 2)
pop = np.array([census2000['Total'].sum(), census2010['Total'].sum()])

s = InterpolatedUnivariateSpline(yr, pop, k = 1)
population = map(round, s(np.linspace(2001, 2013, 13)))

# Group the crimes into categories
crimes_category = crimes_data.groupby(['Year', 'Category']).size()
crimes_category = crimes_category.unstack('Category')
crimes_category['Total'] = crimes_category.sum(axis = 1)

# Normalize by population and number of years
crimes_category = crimes_category.div(population[3:], axis = 0)
nyears = 10
crimes_category = crimes_category/nyears*100000

crimes_proportion = pd.DataFrame()
crimes_category = crimes_category.reindex_axis(['Property', 'Violent', 'Drug', 'Other', 'Total'], axis = 1)
for col in ['Property', 'Violent', 'Drug', 'Other']:
    crimes_proportion[col] = crimes_category[col]/crimes_category['Total']
crimes_proportion.index = crimes_proportion.index.map(str)
crimes_proportion.index.name = 'Year'
```

```python
# Group the crimes into categories
crimes_category = crimes_data.groupby(['Year', 'Category']).size()
crimes_category = crimes_category.unstack('Category')
crimes_category['Total'] = crimes_category.sum(axis = 1)

# Normalize by population and number of years
crimes_category = crimes_category.div(population[3:], axis = 0)
nyears = 10
crimes_category = crimes_category/nyears*100000
crimes_category.index = crimes_category.index.map(str)
crimes_category.index.name = 'Year'
crimes_category.columns.name = None
```

```python
fig = plt.figure(figsize = (12, 4))
ax1 = fig.add_subplot(1, 2, 1)
crimes_category[['Total']].plot(ax = ax1, color = 'red', linewidth = 2)
ax2 = fig.add_subplot(1, 2, 2)
crimes_proportion.plot(ax = ax2, kind = 'bar', stacked = True, secondary_y = True, ylim = [0, 1])
ax2.set_xlabel('Year')
plt.suptitle('Crime rate in Chicago per 100,000 people per year\n', fontsize = 14)
plt.legend(loc='center left', bbox_to_anchor=(1.08, 0.5))
plt.subplots_adjust(right = 0.8)
fig.savefig('../figs/crimes.png')
```

### Which crimes have decreased the most/least between 2001 and 2013?

```python
crimes_type = crimes_data.groupby(['Year', 'Primary Type']).size().unstack('Primary Type').fillna(0)
crimes_type['TOTAL'] = crimes_type.sum(axis = 1)

# Considering the individual crime types
# Combine assault and battery
crimes_type['ASSAULT/BATTERY'] = crimes_type['ASSAULT'] + crimes_type['BATTERY']
crimes_type.drop(['ASSAULT', 'BATTERY'], axis = 1, inplace = True)

# Combine criminal damage and criminal trespass
crimes_type['CRIMINAL DAMAGE/TRESPASS'] = crimes_type['CRIMINAL DAMAGE'] + crimes_type['CRIMINAL TRESPASS']
crimes_type.drop(['CRIMINAL DAMAGE', 'CRIMINAL TRESPASS'], axis = 1, inplace = True)

# Rename deceptive practice fraud
crimes_type.rename(columns = {'DECEPTIVE PRACTICE': 'FRAUD'}, inplace = True)

# Drop crime types with less than 10000 occurrences except homicides, and other offense
drop = ((crimes_type < 10000).all(axis = 0)) & ~(crimes_type.columns == 'HOMICIDE') | \
       (crimes_type.columns == 'OTHER OFFENSE')
crimes_type.drop(crimes_type.columns[drop], axis = 1, inplace = True)

# Normalize by population and number of years and sort by order of prevalence
crimes_type = crimes_type.div(population[3:], axis = 0)
crimes_type = crimes_type/nyears*100000

crimes_type.ix[2004].argsort()
crimes_type = crimes_type.reindex_axis(crimes_type.sum().order(ascending = False).index, axis = 1)

crimes_type_n = crimes_type.copy(deep = True)

# Normalize to year 2004
norm = crimes_type.ix[2004]
crimes_type_n = crimes_type_n.div(norm, axis = 1)
 
# Change in crime between 2004 and 2013 
change = crimes_type.ix[2004, 1:]
current = crimes_type.ix[2013, 1:]

change = (current - change)/change*100
change = change[~change.isnull()]
change.index.name = None

change_category = crimes_category.ix['2004', ]
current_category = crimes_category.ix['2013', ]

change_category = (current_category - change_category)/change_category*100
change_category = change_category[~change_category.isnull()]
```

```python
fig = plt.figure(figsize = (12, 6.5))
ax1 = fig.add_subplot(1, 2, 1)
change.order().plot(ax = ax1, kind = 'bar')
ax1.set_xlabel = None
ax1.set_ylabel = '%'
ax2 = fig.add_subplot(1, 2, 2)
change_category.plot(ax = ax2, kind = 'bar')
ax2.set_ylabel = '%'
plt.suptitle('Percent change in crime rate between 2004 and 2013\n', fontsize = 14)
plt.subplots_adjust(bottom = 0.375)
fig.savefig('../figs/crimes_change.png')
```

Most crime types show a broad decline over the years. However, different crimes show different patterns of occurrence. Burglary shows a broad increase between 2007 and 2011.

Overall crime rate tends to increase with decreasing income, but there are some areas with high income (peaks in the per capita income curve) that have high crimes rates. The types of crimes that occur gives some idea why.

`*`The data for the per capita income of Chicago community areas was downloaded from the Chicago data portal.

The proportion of property crimes increases as the income goes up, while the proportion of drug crimes increases as the income goes down. The proportion of violent crimes tends to stay mostly constant with income.

```python
# Change TMAX and TMIN to Fahrenheit
weather = pd.read_csv('../data/weather_from_2001_to_2014.csv', usecols = [2, 6, 7], 
                      parse_dates = ['DATE'], na_values = [-9999])

# Change DATE(Timestamp) to Date(Datetime.date)
weather['Date'] = weather['DATE'].map(lambda x: x.date())

weather['Tmax_F'] = weather['TMAX'].astype(float)*0.9/5 + 32
weather['Tmin_F'] = weather['TMIN'].astype(float)*0.9/5 + 32
weather.drop(['DATE', 'TMAX', 'TMIN'], axis = 1, inplace = True)

crimes_grouped = pd.merge(crimes_grouped, weather, on = 'Date', how = 'left').fillna(method = 'ffill')
```

```python
# Set up a test dataframe for the week of Aug 4 2014.
test = pd.DataFrame({'key': [0]*21, 'DateTime': pd.date_range(start, periods = 21, freq = '8H')})
test['Date'] = test['DateTime'].map(lambda x: x.date())
test['Day of Week'] = test['DateTime'].map(lambda x: days[x.weekday()])

times = [0, 8, 16, 24]
shifts = ['0', '1', '2'] 
test['Shift'] = pd.cut(test['DateTime'].map(lambda x: x.hour), times, right = False, labels = shifts)

test = pd.merge(test, pd.DataFrame({'key': [0]*77, 'Community Area': range(1, 78)}), 
                   how = 'outer').drop('key', axis = 1)

# Merge with actual crimes
crimes_grouped_test = crimes_grouped[(crimes_grouped['Date'] >= start.date()) & (crimes_grouped['Date'] <= end.date())]
test = pd.merge(test, crimes_grouped_test[['Date', 'Day of Week', 'Shift', 'Community Area','Crimes']], 
                how = 'left', on = ['Date', 'Day of Week', 'Shift', 'Community Area']).fillna(0)

test['Year'] = test['DateTime'].map(lambda x: x.year)

test['CA, Time'] = 'CA' + test['Community Area'].map(str) + ', ' + test['Day of Week'] + \
                             ', Shift' + test['Shift']
    
#test = pd.merge(test, socioeconomic[['Community Area', 'Per Capita Income']], on = 'Community Area', how = 'outer')

test = pd.merge(test, weather, on = 'Date', how = 'left').fillna(method = 'ffill')
```

```python
train = crimes_grouped[(crimes_grouped['Date'] >= trainstart.date()) & (crimes_grouped['Date'] <= trainend.date())]
                      
v = DictVectorizer()
arr_train = v.fit_transform(train[['CA, Time']].T.to_dict().values())
#X_train = scipy.sparse.hstack([arr_train, train[['Year', 'Per Capita Income', 'Tmax_F', 'Tmin_F']].values])
X_train = scipy.sparse.hstack([arr_train, train[['Year', 'Tmax_F', 'Tmin_F']].values])
y_train = np.log(train['Crimes'].values + 1)

arr_test = v.fit_transform(test[['CA, Time']].T.to_dict().values())
#X_test = scipy.sparse.hstack([arr_test, test[['Year', 'Per Capita Income', 'Tmax_F', 'Tmin_F']].values])
X_test = scipy.sparse.hstack([arr_test, test[['Year', 'Tmax_F', 'Tmin_F']].values])
y_test = test['Crimes'].values
```
