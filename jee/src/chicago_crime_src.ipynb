## Where did the data come from?

```python
import pandas as pd

crimes = pd.read_csv('../data/crimes_2001_to_present.csv', usecols = [1, 2, 5, 8, 13, 17, 19, 20])
```

```python
# Parse the date field
crimes['DateTime'] = crimes['Date'].map(pd.to_datetime)
crimes = crimes.drop('Date', axis = 1)

# Only consider data from 2004 (the last 10 years)
crimes = crimes[crimes['DateTime'] >= dt.datetime(2004, 1, 1)]

# Clean up community area field
crimes['Community Area'] = crimes['Community Area'].map(lambda x: 
                                                        int(x) if not (pd.isnull(x) or x == ' ') else 0)
crimes = crimes[crimes['Community Area'] != 0]
```

The crimes data was downloaded from the Chicago data portal, covering incidences of crime from 2001, and contains over 5.5 million records.

```python
import pandas as pd

# The demographics table was obtained by analyzing the 2010 census data for the city of Chicago.
# census data for age and sex for different races
demo = {}
for d in ['P12', 'P12I', 'P12H', 'P12B', 'P12D']:
    df = pd.read_csv('../data/DEC_10_SF1_' + d + '_with_ann.csv', header = 1)
    demo[d] = df[df['Geography'] == 'Chicago city, Illinois']
    
group_names = ['Under 5', '5 to 14', '15 to 24', '25 to 34', '35 to 44', '45 to 54',
               '55 to 64', 'Over 65', 'Total']

# Regroup the census data into the above age groups
def group_ages(s):
    age_groups = []
    age_groups.append(s[1])
    age_groups.append(s[2:4].sum())
    age_groups.append(s[4:9].sum())
    age_groups.append(s[9:11].sum())
    age_groups.append(s[11:13].sum())
    age_groups.append(s[13:15].sum())
    age_groups.append(s[15:18].sum())
    age_groups.append(s[18:].sum())
    age_groups.append(s[0])
    return age_groups

P12_R = pd.DataFrame({'Male': group_ages(demo['P12'].ix[:, 4:28].stack()), 
                      'Female': group_ages(demo['P12'].ix[:, 28:].stack())}, index = group_names)
P12IH_R = pd.DataFrame({'Male': [sum(x) for x in zip(group_ages(demo['P12I'].ix[:, 4:28].stack()), 
                                                     group_ages(demo['P12H'].ix[:, 4:28].stack()))], 
                        'Female': [sum(x) for x in zip(group_ages(demo['P12I'].ix[:, 28:].stack()), 
                                                       group_ages(demo['P12H'].ix[:, 28:].stack()))]}, 
                       index = group_names)
P12B_R = pd.DataFrame({'Male': group_ages(demo['P12B'].ix[:, 4:28].stack()), 
                       'Female': group_ages(demo['P12B'].ix[:, 28:].stack())}, index = group_names)
P12D_R = pd.DataFrame({'Male': group_ages(demo['P12D'].ix[:, 4:28].stack()), 
                       'Female': group_ages(demo['P12B'].ix[:, 28:].stack())}, index = group_names)

demographics = pd.concat([P12_R, P12IH_R, P12B_R, P12D_R], axis = 0, 
                         keys = ['ALL', 'WHITE', 'AFAM', 'ASIAN'])
```

```python
from pandas.io.excel import read_excel

# The demographics by community area data was downloaded from 
# http://robparal.blogspot.com/2012/05/hard-to-find-census-data-on-chicago.html
names = ['Community Area', 'Neighborhood', 'NHW', 'AFAM', 'ASIAN', 'HISP', 'Total']

census = {}
census[2000] = read_excel('../data/Race_by_Community_Area_2000_2010.xlsx', skiprows = [0, 1, 2, 3, 4], 
                           parse_cols = 'A:D, F, H, J', names = names, header = None, skip_footer = 10,
                           index_col = 'Community Area')
census[2010] = read_excel('../data/Race_by_Community_Area_2000_2010.xlsx', skiprows = [0, 1, 2, 3, 4], 
                           parse_cols = 'L:O, Q, S, U', names = names, header = None, skip_footer = 10,
                           index_col = 'Community Area')

comm_area = pd.read_csv('../data/CommAreas.csv', usecols = [4, 5, 7], index_col = 'AREA_NUMBE').sort_index()

for _, v in census.iteritems():
    # sum NHW and HISP columns into WHITE
    v['White'] = v['NHW'] + v['HISP']

    # include population density field
    # convert sq ft into sq mi
    v['Population Density'] = v['Total'].astype(float)/comm_area['SHAPE_AREA']*27878400
    v.drop('TOTAL', axis = 0, inplace = True)
```

The demographics table was obtained by analyzing the 2010 census data for the city of Chicago. It tabulates the population grouped into age brackets and sex. The tables referenced are DEC_10_SF1_P12_with_ann.csv, DEC_10_SF1_P12B_with_ann.csv, DEC_10_SF1_P12D_with_ann.csv, DEC_10_SF1_P12H_with_ann.csv, and DEC_10_SF1_P12I_with_ann.csv.

The census data that gives the demographics of Chicago community areas was downloaded from http://robparal.blogspot.com/2012/05/hard-to-find-census-data-on-chicago.html.

The total land area of community areas was downloaded from the chicago data portal.

```python
import dateutil.parser

def date_parser(date_str, time_str):
    if pd.isnull(time_str):
        time_str = ' '
    return dateutil.parser.parse(' '.join([date_str, time_str]))

names = ['Address', 'Date', 'Time', 'Location', 'Neighborhood', 'Age', 'Gender', 
         'Race', 'Name', 'Cause', 'Story URL', 'Case Number', 'Charges and Trials']

hom = {}
for y in range(2007, 2014):
    hom[y] = pd.read_csv('../data/homicides_' + str(y) + '.csv', parse_dates= [['Date', 'Time']],
                          date_parser = date_parser, skiprows = 1, names = names)

homicides = pd.concat(hom.values(), axis = 0)
homicides['Year'] = homicides['Date_Time'].map(lambda x: x.year)

# process 'Age' field into a numeric 'Age_R' field.
def get_age(age_str):
    if pd.isnull(age_str): return None
    if age_str == 'Newborn': 
        return 0
    elif ('day' in age_str) or ('week' in age_str) or ('month' in age_str):
        return 0
    elif '20s' in age_str:
        return 25
    elif '30s' in age_str:
        return 35
    else:
        return int(age_str)
    
homicides['Age_R'] = homicides['Age'].map(get_age)

age_groups = ['Under 5', '5 to 14', '15 to 24', '25 to 34', '35 to 44', '45 to 54',
              '55 to 64', 'Over 65']
ages = [0, 5, 15, 25, 35, 45, 55, 65, 100]

homicides['Age Group'] = pd.cut(homicides['Age_R'], ages, right = False, labels = age_groups)
homicides = homicides.drop('Age_R', axis = 1)
```

The homicides data were downloaded from the redeye Chicago homicide tracker website, http://homicides.redeyechicago.com/. The dataset includes 100 more homicides over the period 2007-2013 than the dataset from the Chicago data portal, and includes demographic information on the victims.

`*`In aggregating the murders into victim's age groups, there are 64 datapoints with missing ages. Those indicated 'unknown: 20s' (3 datapoints) and 'unknown: 30s' (1 datapoint) were reclassified as being of age 25 and 35, respectively.

## Analyzing Murders in Chicago

### Total murders each year per 100,000 people in Chicago.

```python
import numpy as np
from scipy.interpolate import InterpolatedUnivariateSpline 

# Interpolate the population linearly between 2000 and 2010
yr = np.linspace(2001, 2010, 2)
pop = np.array([census[2000]['Total'].sum(), census[2010]['Total'].sum()])

s = InterpolatedUnivariateSpline(yr, pop, k = 1)
population = map(round, s(np.linspace(2001, 2013, 13)))

# The number of murders per 100000 people per year in Chicago
homicides_by_year = homicides.groupby('Year').size()
homicides_by_year = homicides_by_year/population[6:]*100000
```

```python
homicides_by_year.plot(kind = 'bar', figsize = (6, 5))
plt.title('Murders per 100,000 people in Chicago\n', fontsize = 14)
plt.subplots_adjust(top = 0.85, bottom = 0.15)
plt.savefig('../figs/murders.png')
```

Unlike the overall crime rate in Chicago which shows a steady decrease over the years, the rate of murder is broadly consistent  with peaks in 2008 and 2012.

`*`The population of Chicago for each year was obtained by linearly interpolating between the populations at 2000 and 2010.

### Murder rates by race, gender and age

```python
# Clean up the race field
homicides['Race'] = homicides['Race'].str.rstrip().str.upper()
homicides.replace('BLACK', 'AFAM', inplace = True)

# Consider the year 2013
homicides13 = homicides[homicides['Year']==2013]

# The number of murders by race of victim
race = homicides13.groupby(['Race']).size()

pop = pd.Series([demographics.ix[('ASIAN', 'Total'), ].sum(),
                 demographics.ix[('AFAM', 'Total'), ].sum(),
                 demographics.ix[('WHITE', 'Total'), ].sum()], index = ['ASIAN', 'AFAM', 'WHITE'])

race = race/pop*100000
race.index.name = None

# The number of murders by race and gender of victim
gender = homicides13.groupby(['Race', 'Gender']).size()

pop_gender = pd.concat([demographics.ix[('ASIAN', 'Total'), ], 
                        demographics.ix[('AFAM', 'Total'), ],
                        demographics.ix[('WHITE', 'Total'), ]], keys = ['ASIAN', 'AFAM', 'WHITE'])
pop_gender.index.names = ['Race', 'Gender']

gender = (gender/pop_gender*100000).unstack('Gender')
gender.columns.name = None
gender.index.name = None

# The number of murders by race and age of victim
age = homicides13.groupby(['Race', 'Age Group']).size()

pop_age = pd.concat([demographics.ix[('ASIAN', ), ].sum(1)[:-1],
                     demographics.ix[('AFAM', ), ].sum(1)[:-1],
                     demographics.ix[('WHITE', ), ].sum(1)[:-1]], keys = ['ASIAN', 'AFAM', 'WHITE'])
pop_age.index.names = ['Race', 'Age Group']

age = (age/pop_age*100000).fillna(0).unstack()
age.columns.name = None
cols = ['Under 5', '5 to 14', '15 to 24', '25 to 34', '35 to 44', '45 to 54', '55 to 64', 'Over 65']
age = age[cols]
age.index.name = None

# Coverage of murders in the media
media = (pd.isnull(homicides['Story URL'])) | (pd.isnull(homicides['Charges and Trials']))
coverage = homicides[~media].groupby(['Race']).size()/race
coverage.index.name = None
```

```python
fig = plt.figure(figsize = (10, 8))

ax1 = plt.subplot2grid((2, 2), (0, 0))
ax2 = plt.subplot2grid((2, 2), (0, 1))
ax3 = plt.subplot2grid((2, 2), (1, 0), colspan = 2)
race.plot(ax = ax1, kind = 'bar')
gender.plot(ax = ax2, kind = 'bar')
age.plot(ax = ax3, kind = 'bar')
ax3.legend(loc='best')
plt.subplots_adjust(hspace = 0.3)
fig.suptitle('Murders per 100,000 people by race, gender and age, in 2013', fontsize = 14)

plt.savefig('../figs/murders_grouped.png')
```

## Overall Trends in Crime

Violent crimes are are those committed directly against a person: Homicide, Criminal Sexual Assault, Robbery, and Aggravated Assault/Battery. Property crimes are those in which there is no direct threat or harm to a person - Burglary, Theft, Motor Vehicle Theft, Criminal Damage, and Arson.

```python
#crimes_data = crimes.copy()
crimes_data = crimes[crimes['Year'] < 2014]

# Divide the crimes into categories of violent, property, drug, and other.
Violent_Crime = ['HOMICIDE', 'CRIM SEXUAL ASSAULT', 'ASSAULT', 'BATTERY', 'ROBBERY']
Property_Crime = ['THEFT', 'BURGLARY', 'CRIMINAL DAMAGE', 'MOTOR VEHICLE THEFT', 'ARSON']
Drug_Crime = ['NARCOTICS']

def get_category(type):
    if type in Violent_Crime:
        return 'VIOLENT'
    elif type in Property_Crime:
        return 'PROPERTY'
    elif type in Drug_Crime:
        return 'DRUG'
    else:
        return 'OTHER'
    
crimes_data['Category'] = crimes_data['Primary Type'].apply(get_category)

# The population of each year is interpolated between the populations of 2000 and 2010
from scipy.interpolate import InterpolatedUnivariateSpline 

yr = np.linspace(2001, 2010, 2)
pop = np.array([census[2000]['Total'].sum(), census[2010]['Total'].sum()])

s = InterpolatedUnivariateSpline(yr, pop, k = 1)
population = map(round, s(np.linspace(2001, 2013, 13)))

# Group the crimes into categories
crimes_category = crimes_data.groupby(['Year', 'Category']).size()
crimes_category = crimes_category.unstack('Category')
crimes_category['TOTAL'] = crimes_category.sum(axis = 1)

# Normalize by population
crimes_category = crimes_category.div(population[3:], axis = 0)
crimes_category.index = crimes_category.index.map(str)
crimes_category.index.name = 'Year'
crimes_category.columns.name = None

# The proportions of property, violent and drug crimes
crimes_proportion = pd.DataFrame()
for col in ['PROPERTY', 'VIOLENT', 'DRUG', 'OTHER']:
    crimes_proportion[col] = crimes_category[col]/crimes_category['TOTAL']
crimes_proportion.index = crimes_proportion.index.map(str)
crimes_proportion.index.name = 'Year'
```

```python
fig = plt.figure(figsize = (12, 4))
ax1 = fig.add_subplot(1, 2, 1)
crimes_category[['TOTAL']].plot(ax = ax1, color = 'red', linewidth = 2)
ax2 = fig.add_subplot(1, 2, 2)
crimes_proportion.plot(ax = ax2, kind = 'bar', stacked = True, secondary_y = True, ylim = [0, 1])

ax1.set_ylabel('Crime rate per capita')
ax1.legend().set_visible(False)
ax2.set_xlabel('Year')
ax1.set_title('Crime rate in Chicago\n', fontsize = 14)
ax2.set_title('Proportion of property, violent, and drug crimes\n', fontsize = 14)
plt.legend(loc = 'center right', bbox_to_anchor=(1.52, 0.5))
plt.subplots_adjust(top = 0.85, right = 0.8)
fig.savefig('../figs/crimes.png')
```

### Which crimes have decreased the most/least between 2001 and 2013?

```python
crimes_type = crimes_data.groupby(['Year', 'Primary Type']).size().unstack('Primary Type').fillna(0)
crimes_type['TOTAL'] = crimes_type.sum(axis = 1)

# Combine assault and battery
crimes_type['ASSAULT/BATTERY'] = crimes_type['ASSAULT'] + crimes_type['BATTERY']
crimes_type.drop(['ASSAULT', 'BATTERY'], axis = 1, inplace = True)

# Combine criminal damage and criminal trespass
crimes_type['CRIMINAL DAMAGE/TRESPASS'] = crimes_type['CRIMINAL DAMAGE'] + crimes_type['CRIMINAL TRESPASS']
crimes_type.drop(['CRIMINAL DAMAGE', 'CRIMINAL TRESPASS'], axis = 1, inplace = True)

# Rename deceptive practice fraud
crimes_type.rename(columns = {'DECEPTIVE PRACTICE': 'FRAUD'}, inplace = True)

# Drop crime types with less than 10000 occurrences in every year except homicides, and other offense
drop = ((crimes_type < 10000).all(axis = 0)) & ~(crimes_type.columns == 'HOMICIDE') | \
       (crimes_type.columns == 'OTHER OFFENSE')
crimes_type.drop(crimes_type.columns[drop], axis = 1, inplace = True)

# Normalize by population
crimes_type = crimes_type.div(population[3:], axis = 0)

crimes_type.ix[2004].argsort()
crimes_type = crimes_type.reindex_axis(crimes_type.sum().order(ascending = False).index, axis = 1)

# Change in crime between 2004 and 2013 
initial = crimes_type.ix[2004, 1:]
current = crimes_type.ix[2013, 1:]
change = (current - initial)/initial*100
change = change[~change.isnull()]
change.index.name = None

initial_cat = crimes_category.ix['2004', ]
current_cat = crimes_category.ix['2013', ]
change_cat = (current_cat - initial_cat)/initial_cat*100
change_cat = change_cat[~change_cat.isnull()]
```

Assault and battery have been combined in the following, although they are separate crime types in the dataset. An assault is an attempt or threat to injure another person, while a battery would be actually contacting another person in a harmful or offensive manner. Similarly, criminal damage and criminal trespass have also been combined.

Deceptive practice has been renamed Fraud.

```python
fig = plt.figure(figsize = (12, 6.5))
ax1 = fig.add_subplot(1, 2, 1)
change.order().plot(ax = ax1, kind = 'bar')
ax1.set_xlabel = None
ax1.set_ylabel('%')
ax2 = fig.add_subplot(1, 2, 2)
change_cat.plot(ax = ax2, kind = 'bar')
ax2.set_ylabel('%')
plt.suptitle('Percent change in crime rate between 2004 and 2013\n', fontsize = 14)
plt.subplots_adjust(bottom = 0.375)
fig.savefig('../figs/crimes_change.png')
```

Drug and violent crime rates have decreased the most while property crimes rates decresed the least between 2004 and 2013.

### Training and Cross-Validation

A linear model based on the community, day of week and time of day (8 hour periods starting from 00:00), and the maximum and minimum temperatures of that day.

A training/cv set was taken bewteen Jan. 2001 and Dec. 2013. Two cross-validation methods were used: first, for a training period of 2 years and a cross-validation period of 1 week immediately following the training period, with a moving window of 1 month, the Ridge parameter alpha that resulted in the best average Spearman's rank correlation coefficient between the predicted and actual number of crimes was found to be close to 0.

Second, using a log-regression (based on the previous result of alpha = 0), the training period was varied between 1 and 5 years, with the window moving forward by 1 month each time to find that the training period that resulted in the best average Spearman's rank correlation coefficient between the predicted and actual number of crimes was 1 year.

```python
from pandas.tseries.offsets import DateOffset

# Set up a test dataframe for a week from the start date.
def setup_test(start):
    test = pd.DataFrame({'key': [0]*21, 'DateTime': pd.date_range(start, periods = 21, freq = '8H')})
    test['Date'] = test['DateTime'].map(lambda x: x.date())
    test['Day of Week'] = test['DateTime'].map(lambda x: days[x.weekday()])

    times = [0, 8, 16, 24]
    shifts = ['0', '1', '2'] 
    test['Shift'] = pd.cut(test['DateTime'].map(lambda x: x.hour), times, right = False, labels = shifts)

    test = pd.merge(test, pd.DataFrame({'key': [0]*77, 'Community Area': range(1, 78)}), 
                    how = 'outer').drop('key', axis = 1)

    # Merge with actual crimes
    end = start + dt.timedelta(6)
    crimes_grouped_test = crimes_grouped[(crimes_grouped['Date'] >= start.date()) & 
                                         (crimes_grouped['Date'] <= end.date())]
    join_cols = ['Date', 'Day of Week', 'Shift', 'Community Area']
    test = pd.merge(test, crimes_grouped_test[join_cols + ['Crimes']], how = 'left', on = join_cols).fillna(0)
    test['Year'] = test['DateTime'].map(lambda x: x.year)
    test['CA, Time'] = 'CA' + test['Community Area'].map(str) + ', ' + test['Day of Week'] + \
                                   ', Shift' + test['Shift']
    test = pd.merge(test, weather, left_on = 'Date', right_index = True)

    return test
```

```python
from sklearn.feature_extraction import DictVectorizer
from sklearn import linear_model
from scipy.stats import spearmanr

# Try a range of alpha values
alphas = np.logspace(-4, 2, 10)

# Use 2 years' data for training starting from Jan 1st 2004
nyears = 2
start = dt.datetime(2004, 1, 1)
end = dt.datetime(2013, 12, 31)

cors = []
for alpha in alphas:
    cor = []
    trainstart = start
    while trainstart + dt.timedelta(365*nyears + 7) < end:
        trainend = trainstart + dt.timedelta(365*nyears)
        teststart = trainend + dt.timedelta(1)
        testend = teststart + dt.timedelta(6)
        
        test = setup_test(teststart)
        
        train = crimes_grouped[(crimes_grouped['Date'] >= trainstart.date()) & 
                               (crimes_grouped['Date'] <= trainend.date())]
                      
        v = DictVectorizer()
        arr_train = v.fit_transform(train[['CA, Time']].T.to_dict().values())
        X_train = scipy.sparse.hstack([arr_train, train[['Year', 'Tmax_F', 'Tmin_F']].values])
        y_train = np.log(train['Crimes'].values + 1)

        arr_test = v.fit_transform(test[['CA, Time']].T.to_dict().values())
        X_test = scipy.sparse.hstack([arr_test, test[['Year', 'Tmax_F', 'Tmin_F']].values])
        y_test = test['Crimes'].values
        
        clf = linear_model.Ridge(alpha = alpha)
        clf.fit(X_train, y_train)
        y_pred = np.exp(clf.predict(X_test)) - 1
        cor.append(spearmanr(y_pred, y_test)[0])
        
        trainstart = trainstart + dt.timedelta(30)
        
    cors.append(np.mean(cor))

ridge_models = pd.DataFrame({'alpha': alphas, 'spearmanr': cors})
ridge_models.plot(x = 'alpha', y  ='spearmanr')
plt.ylabel('Spearman Rank Correlation Coefficient')
```

```python
crimes_grouped.sort_index(inplace = True)

alphas = np.logspace(-3., 2, 16)
start = dt.datetime(2001, 1, 1)
end = dt.datetime(2013, 12, 31)
nyears = 2
cors = []
for alpha in alphas:
    cor = []
    trainstart = start
    while trainstart + dt.timedelta(365*(nyears + 1) + 31) < end:
        
        trainend = trainstart + dt.timedelta(365*(nyears + 1))
        teststart = trainend + dt.timedelta(1)
        testend = teststart + dt.timedelta(30)

        X_train = crimes_grouped.ix[trainstart.date():trainend.date(), 2:94]
        y_train = crimes_grouped.ix[trainstart.date():trainend.date(), 'Crimes']

        X_test = crimes_grouped.ix[teststart.date():testend.date(), 2:94]
        y_test = crimes_grouped.ix[teststart.date():testend.date(), 'Crimes']
 
        clf = linear_model.Ridge(alpha = alpha)

        cor.append(scipy.stats.spearmanr(clf.fit(X_train, y_train).predict(X_test), y_test)[0])

        trainstart = trainstart + dt.timedelta(30)
              
    cors.append(np.mean(cor))
    
ridge_models = pd.DataFrame({'alpha':alphas, 'spearmanr': cors})
ridge_models.plot(x='alpha', y='spearmanr')
```

### Training and Predicting Crime Occurrence

```python
from sklearn.feature_extraction import DictVectorizer
from pandas.tseries.offsets import DateOffset

# Predict crime occurrence for the first week of August 2014 
start = dt.datetime(2014, 8, 4)
end = start + DateOffset(6)

# Use the previous year's data for training 
trainstart = dt.datetime(2013, 8, 4)
trainend = dt.datetime(2014, 8, 3)
```

```python
# Set up a test dataframe for the week of Aug 4 2014.
test = pd.DataFrame({'key': [0]*21, 'DateTime': pd.date_range(start, periods = 21, freq = '8H')})
test['Date'] = test['DateTime'].map(lambda x: x.date())
test['Day of Week'] = test['DateTime'].map(lambda x: days[x.weekday()])

times = [0, 8, 16, 24]
shifts = ['0', '1', '2'] 
test['Shift'] = pd.cut(test['DateTime'].map(lambda x: x.hour), times, right = False, labels = shifts)

test = pd.merge(test, pd.DataFrame({'key': [0]*77, 'Community Area': range(1, 78)}), 
                how = 'outer').drop('key', axis = 1)

# Merge with actual crimes
crimes_grouped_test = crimes_grouped[(crimes_grouped['Date'] >= start.date()) & (crimes_grouped['Date'] <= end.date())]
test = pd.merge(test, crimes_grouped_test[['Date', 'Day of Week', 'Shift', 'Community Area','Crimes']], 
                how = 'left', on = ['Date', 'Day of Week', 'Shift', 'Community Area']).fillna(0)
test['Year'] = test['DateTime'].map(lambda x: x.year)
test['CA, Time'] = 'CA' + test['Community Area'].map(str) + ', ' + test['Day of Week'] + \
                             ', Shift' + test['Shift']
test = pd.merge(test, weather, on = 'Date', how = 'left').fillna(method = 'ffill')
```

```python
train = crimes_grouped[(crimes_grouped['Date'] >= trainstart.date()) & 
                       (crimes_grouped['Date'] <= trainend.date())]
                      
v = DictVectorizer()
arr_train = v.fit_transform(train[['CA, Time']].T.to_dict().values())
X_train = scipy.sparse.hstack([arr_train, train[['Year', 'Tmax_F', 'Tmin_F']].values])
y_train = np.log(train['Crimes'].values + 1)

arr_test = v.fit_transform(test[['CA, Time']].T.to_dict().values())
X_test = scipy.sparse.hstack([arr_test, test[['Year', 'Tmax_F', 'Tmin_F']].values])
y_test = test['Crimes'].values
```

```python
from sklearn import linear_model

clf = linear_model.LinearRegression()
clf.fit(X_train, y_train)
y_pred = np.exp(clf.predict(X_test)) - 1

test['Predicted'] = y_pred
```

```python
test.plot(x = 'Crimes', y = 'Predicted', kind = 'scatter', figsize = (6.5, 4.5),
          xlim = (0, 42), ylim = (0, 30))
plt.title('Predicted And Actual Crimes During 8/4/2014 - 8/10/2014\n', fontsize = 14)
plt.subplots_adjust(top = 0.85)
plt.savefig('../figs/predict.png')
```

```python
test = pd.merge(test, census[2010][['Total']], left_on = 'Community Area', right_index = True)

total_pop = census[2010]['Total'].sum().astype(float)
num = 10
predicted = []; predicted_pop = []
actual = []; actual_pop = []
random = []; random_pop = []
for d in range(7):
    for s in range(3):
        test_ds = test[(test['Day of Week'] == days[d]) & (test['Shift'] == str(s))]
        total = test_ds['Crimes'].sum().astype(float)
        predicted.append(test_ds.sort('Predicted')[-num:]['Crimes'].sum()/total)
        predicted_pop.append(test_ds.sort('Predicted')[-num:]['Total'].sum()/total_pop)
        actual.append(test_ds.sort('Crimes')[-num:]['Crimes'].sum()/total)
        actual_pop.append(test_ds.sort('Crimes')[-num:]['Total'].sum()/total_pop)
        rand = np.random.random_integers(0, 76, 10)
        random.append(test_ds.iloc[rand]['Crimes'].sum()/total)
        random_pop.append(test_ds.iloc[rand]['Total'].sum()/total_pop)
        
summary = pd.DataFrame({'Actual Worst 10': actual, 'Predicted Worst 10': predicted, 'Random 10': random})
summary_pop = pd.DataFrame({'Actual': actual_pop, 'Predicted': predicted_pop, 'Random': random_pop})
```

```python
num = 10
predicted_shift = []
actual_shift = []
random_shift = []
for c in range(1, 78):
    test_ds = test[test['Community Area'] == c]
    total = test_ds['Crimes'].sum().astype(float)
    predicted_shift.append(test_ds.sort('Predicted')[-num:]['Crimes'].sum()/total)
    actual_shift.append(test_ds.sort('Crimes')[-num:]['Crimes'].sum()/total)
    rand = np.random.random_integers(0, 20, 10)
    random_shift.append(test_ds.iloc[rand]['Crimes'].sum()/total)
        
summary_shift = pd.DataFrame({'Actual Worst 10': actual_shift, 'Predicted Worst 10': predicted_shift, 
                              'Random 10': random_shift})
```
