```python
# Get rid of neutral sentences. DON'T DO THIS AT THE START. TRAIN ON THE NON-NEUTRAL. TEST ON EVRYTHING
df = df[df['Sentiment'] != 2]
```

```python
#Stemming words to avoid overcounting derivatives of words
#lemmatizer stems words to actual words which is nicer
#On the other hand lemmatizer sucks. Get a new one.

lemmatizer = nltk.stem.wordnet.WordNetLemmatizer()

#Stems words of a sentence and puts them back together
def Stemm(sentence):
    return " ".join(map(lemmatizer.lemmatize, sentence.split(" ")))

print Stemm("Bob bobs for as was apples. Bob's your uncle")

stemmer = nltk.stem.SnowballStemmer("english", ignore_stopwords=True)

print stemmer.stem("The quick brown foxes jumped over the carries carrying carried lazy laziness dog. ") 


#df['Phrase']=df['Phrase'].apply(Stemm)
df['Phrase']=df['Phrase'].apply(stemmer.stem)
```

```python
#def negation(sentence): This changes the word next to a negative to a negation of the word

neg_words={'not':True,"isn't":True,"wasn't":True,"wouldn't":True,"weren't":True,"never":True,"neither":True,"nor":True,
                "couldn't":True,'no':True,"hadn't":True,"shouldn't":True,'no':True,'no':True,'without':True}
def negate(sentence):
    words=sentence.split(' ')
    for i in range(len(words)):
        if words[i] in neg_words:
            try: words[i+1]='not_' + words[i+1]
            except IndexError: 1
    words=[word for word in words if not word in neg_words]
    return ' '.join(words)
print negate("This film is not good")
print negate("This film is not any good")
print negate("This movie is neither sweet nor funny, neither romantic nor comedic")

#This is a different negator that negates all the words (to a point of punctuation) following a negative word.
import re
def negate2(sentence):
    phrases=re.split('; |, |: |\.',sentence)
    new_phrases=[]
    for phrase in phrases:
        words=phrase.split(' ')
        for i in range(len(words)):
            if words[i] in neg_words:
                words[i+1:]=map(lambda x:'not_'+x,words[i+1:])
        words=[word for word in words if not word in neg_words]
        new_phrase=' '.join(words)
        new_phrases.append(new_phrase)
    return ' '.join(new_phrases)    
print negate2("This film is not good")
print negate2("This film is not any good")
```

```python
df['Phrase']=df['Phrase'].apply(negate2)
```

```python
#This is a vectorizer. It takes in a bunch of sentences and returns a sparse matrix whose rows represent
#the sentences as vectors.

from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(min_df=1,max_df=9000,ngram_range=(1,2))

#examplex=vectorizer.fit_transform(['A boy', 'A girl'])
#examplex.toarray()
```

```python
#This vectorizes the data and centers the sentiment around 0. Don't do this when you put things in a pipeline.
X=vectorizer.fit_transform(df['Phrase'][:75000])
y=df['Sentiment'][:75000]-2
```

```python
#Again. Don't do this. Pipeline
X_train=X[0:50000]
y_train=y[0:50000]

X_CV=X[50000:65000]
y_CV=y[50000:65000]

X_test=X[65000:]
y_test=y[65000:]

#X_test,y_test
```

```python
#Sigmoid function that normalizes ratings centered at 0 to a 0-4 rating
def g(x):
    return 4/(1+np.exp((-x)))

from sklearn import linear_model

regr=linear_model.Ridge(alpha=5)
regr.fit(X_train,y_train)

#print regr.score(X_test,y_test)
#Error r square

print 'The r square score is ', regr.score(X_CV,y_CV)

Err=(g(regr.predict(X_CV))-2-y_CV)

print 'The standard mean square error is ', math.sqrt(sum([math.fabs(t) for t in Err])/15000)
```
