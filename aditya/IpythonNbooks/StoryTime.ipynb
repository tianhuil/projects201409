## The Project

We look at the task of Sentiment Analysis for movie reviews. That is, we want to learn and predict whether a piece of text is  expressing a positive or negative opinion. 

This exercise is quite general. We are always looking to extract concrete quantifiable information from raw unstructured text. 
For instance one might want to read customer comments, twitter feeds and try to estimate support or interest in various products/ issues. Whether a customer will come back to your business, whether a viewer will watch the next season of your show, whether a person will vote for you are all examples of problems of this sort.

Automating the task of mining text to extract useful information is a really interesting area of study and there are many interesting problems to try and solve.

## The Data

The data we use is scraped from the Rotten Tomatoes API. On Rotten Tomatoes, people write movie reviews and also rate movies 
on various number and letter scales. A sample looks like this:
    
{      "critic": "Kimber Myers",
      "date": "2013-08-15",
      "original_score": "C+",
      "freshness": "fresh",
      "publication": "The Playlist",
      "quote": "Paradoxically absorbing, yet cliched and empty, this power-based New York story is a bit of an entertaining thrill ride, probably far more popcorn-engrossing than it deserves to be...",
      "links": {"review": "http://blogs.indiewire.com/theplaylist/review_limitless_test_the_boundaries_of_bradley_coopers_charm"}
    }

We want to fit a model that will take in a review such as above and predict it's star rating on a scale of 0-4.

### Challenges with the Data

Bad Data: The rating does not directly rate the text of the review, but they both evaluate the same movie. This might be
    the case in a real world situation where it might not be possible to manually tag a large number of training samples.
    Also, pretentious movie critics write needlessly convoluted sentences.
    
Bad Humans: Human agreement on the sentiment being expressed is nowhere near a 100%. Which means a program cannot hope 
    for complete accuracy. Accuracy worsens as there are more categories of sentiment. Also pretentious movie critics.Seriously.

Bad Me: I normalized the ratings from various scales pretty much on a whim. Maybe more effort to understand how people rate 
differntly when they are on different scales/ diffferent publications, different personal baseline ratings could have made a significant improvement. This is an intersting exercise in itself that I did not pursue.

## The Model

We start by treating each sentence as a bag of words without regard to semantic structure. Each sentence can then be written as a 
sparse vector (that is the size of the total vocabulary) where we record whether a particular word occurs in a sentence, (and how 
many times). As a processing step, we stem words and throw out very common words.
                                            
These vectors make the rows of 'X's and the rating associated with them are our 'y's. We use a linear model because there is an 
order relation in ratings 1 star < 2 stars <3 stars etc. (I had less success treating it as a classification problem, and using 
Naive Bayes or SVMs.)

### n-grams

2 grams of  ['I will be back']  are  ['I will',  'will be', 'be back' ] 

To incorporate atleast some of the structure of the sentences, I used n-grams in my vocabulary. I cross validated for the range of ngrams to use and the best fit was using 1 and 2-grams.

### Stemming

Stemming identifies  derivatives of the same word

['Carried ','Carriying','Carries']  are stemmed to  ['carri','carri','carri']

### Negation

To deal with negation we introduce new words to the vocabulary which are negations of existing words. 

'This movie is not very good'  becomes 'This movie is not_very not_good'

### Normalisation/Benchmarks

out of the 0-4 range.

```python
import pickle

dataYelp=pickle.load( open( "BigPickledDataRT.p", "rb" ) )
df=dataYelp
dftrain=df[:70000]
#dftrain=dftrain[dftrain['Rating']!=2]
Xtrain=df['Review']
ytrain=df['Rating']

dftest=df[70000:]
Xtest=dftest['Review']
ytest=dftest['Rating']

from sklearn.feature_extraction.text import CountVectorizer
from sklearn import linear_model
from sklearn.pipeline import Pipeline

Vectorizer=CountVectorizer(min_df=10,max_df=10000,ngram_range=(1,2))
regr=linear_model.Ridge(alpha=5)

Vectorized_regr=Pipeline(steps=[('Vectorizer',Vectorizer),('regr',regr)])

Vectorized_regr.set_params(Vectorizer__min_df=10,regr__alpha=5).fit(Xtrain,ytrain)

print Vectorized_regr.score(Xtest,ytest)
Vectorized_regr.predict(Xtest)

Coeff={}
for word in Vectorizer.vocabulary_.keys():
    Coeff[word]=regr.coef_[Vectorizer.vocabulary_[word]]

#regr.coef_.shape
#vectorizer.vocabulary_

Coefficients_of_words=sorted(Coeff.items(), key =lambda x:Coeff[x[0]])

mos_neg=Coefficients_of_words[:20]
mos_pos=Coefficients_of_words[-20:]
mos_pos,mos_neg
```
