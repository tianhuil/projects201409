### Load Data

```python
execfile('data_organization.py')
dat  = load_data_file('data/LoanStats3a.csv')
datb = load_data_file('data/LoanStats3b.csv')

datc = load_data_file('data/LoanStats3c.csv')
all_data = pd.concat((dat, datb, datc), ignore_index=True)
assert(all(dat.columns == datb.columns))
```

```python
def loan_cashflows(loan):
    t = loan.loan_lifetime
    investment = loan.funded_amnt
    repaid = loan.total_pymnt
    installments = int(t / 30.5)
    return array([-1] + [repaid / investment / installments for _ in range(installments)])

cfs = [loan_cashflows(l) for _, l in all_data.iterrows()]

width = max(map(len, cfs))
cf_table = zeros((len(cfs), width))
for i in range(len(cfs)):
    c = cfs[i]
    cf_table[i][:len(c)] = c
df_cf = pd.DataFrame(cf_table, index=all_data.index)

irrs = amap(irr, cf_table)
irrs[~isfinite(irrs)] = -1.

# df_cf = pd.DataFrame(cf_table, index=all_data.index)
```

```python
def count(x): return len(x.index)

def returns(x):
    return (1 + irr(df_cf.ix[x.index].values.sum(0)))**12 - 1

X = grid_search.best_estimator_.get_params()['union'].transform(df_test.values)
proba = pd.Series(grid_search.best_estimator_.get_params()['clf'].predict_proba(X)[:,1],
                  index=df_test.index)
proba *= df_test.int_rate

cuts = percentile(proba, [99.5, 99, 98, 97, 96, 95, 90, 85, 80, 75, 70, 65, 60, 50, 40, 30, 20])
top_count = [sum(proba > cutoff) for cutoff in cuts]
top_frames = [df_test[proba > cutoff] for cutoff in cuts]
rets = [returns(df) for df in top_frames]
```

```python
def bootstrap_returns(df, boots=100):
    size = len(df.index)
    rets = zeros(boots)
    for i in range(boots):
        rets[i] = returns(df.ix[random.choice(df.index, size, replace=True)])
    return rets

ret_boot_dists = [bootstrap_returns(tf) for tf in top_frames]
```

```python
myp.setfigdefaults()
stds = map(lambda x: std(x) * 2, ret_boot_dists)
sems = map(lambda x: 2 * std(x) / sqrt(len(x)), ret_boot_dists)
rets = map(mean, ret_boot_dists)#[returns(df) for df in top_frames]
gcf().set_size_inches(16,8)

errorbar(top_count, rets, stds, fmt='-o', color='k', ecolor='0.7')
errorbar(top_count, rets, sems, fmt='-o', color=(0,0,0,0), ecolor='k')
gca().set_xscale('log')
zip(cuts, rets, [count(df_test[proba > c]) for c in cuts])
bench = returns(df_test)

plot(gca().get_xlim(), [bench, bench], 'k--')
# gca().set_ylim([0, .15])
xlabel('Number of top loans to use of {}'.format(df_test.shape[0]))
ylabel('IRR')
myp.prettify(gca())
gca().set_xlim([100, 2e4]);
```

### Modeling the conditional distribution of returns given that a loan has not yet failed

We want to know the fraction of money returned for an average loan that has survived at least $D$ days, given the loan is of grade $g$. We can start by estimating how much longer the loan is likely to survive:

$$E(d \;|\; d > D, g) = \sum_d d\;p(d\;|\;d>D,g) = \sum_{d>D} d\; \frac{p(d|g)}{p(d>D|g)}$$

```python
def expected_days_given_failure(df, days_old, grd):
    survived_at_least = df.failed & (df.grade == grd) & (width > days_old)
    return width.ix[survived_at_least[survived_at_least].index].mean()

gcf().set_size_inches(12,6)
x = range(0, 1201, 30)
ys = [[expected_days_given_failure(all_data, d, g) for d in x] for g in 'ABCD']
for y in ys: plot(x, y)
legend('ABCD', loc='best')
plot([0,1200], [0,1200], 'k--')
xlabel('Minimum age of loans')
ylabel('Expected age at failure')
```

```python
def expected_days_given_success(df, days_old, grd):
    survived_at_least = df.completed & (~df.failed) & (df.grade == grd) & (width > days_old)
    return width.ix[survived_at_least[survived_at_least].index].mean()

gcf().set_size_inches(12,6)
x = range(0, 1201, 30)
ys = [[expected_days_given_success(all_data, d, g) for d in x] for g in 'ABCD']
for y in ys: plot(x, y)
legend('ABCD', loc='best')
plot([0,1200], [0,1200], 'k--')
xlabel('Minimum age of loans')
ylabel('Expected age at completion')
```

```python
def failure_rate_given_age(df, days_old, grd):
    survived_at_least = df.completed & (df.grade == grd) & (width > days_old)
    fails = df.failed[survived_at_least]
    rate = fails.mean()
    try:
        arate = (rate + 1./2/len(fails.index))
    except:
        return (nan, nan)
    return (rate, sqrt(arate * (1 - arate) / len(fails.index)))

gcf().set_size_inches(12,6)
x = range(0, 1231, 30)
ys = [zip(*[failure_rate_given_age(all_data, d, g) for d in x]) for g in 'ABCD']
for y, e in ys: errorbar(x, y, yerr=e)
legend('ABCD', loc='best', numpoints=1)
xlabel('Minimum age of loans (days)')
ylabel('Expected failure rate')
ylim([0,.3]);
```
