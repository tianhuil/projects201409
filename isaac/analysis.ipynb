## Classification and Modeling

```python
# Benchmark model

df = select_data(dat, balance=True)

logit = LogisticRegression(class_weight='auto')
params = dict(
    penalty = ('l1', 'l2'),
    C = (.001, .01, .1, 1., 10.),
)
bench_grid = GridSearchCV(logit, params, n_jobs=-1, verbose=1, cv=5)
bench_grid.fit(df.subgrade_code.reshape(-1,1), ~df.failed.reshape(-1))
display(bench_grid.best_score_)
display(bench_grid.best_params_)
```

```python
display(sum(datb.total_acc.isnull()))
```

```python
# Text-based model
execfile('models.py')

text_model = Pipeline([
    ('sel', ColumnSelectTransformer(columns=df.columns.get_loc('desc'))),
    ('vect', CountVectorizer()),
    ('tfidf', TfidfTransformer()),
    ('kbest', SelectKBest(k=1000)),
])

linear_vars = ColumnSelectTransformer(astype=float,
    columns=column_locs(df, ['subgrade_code', 'annual_inc', 'total_acc', 'revol_bal']))
category_vars = ColumnSelectTransformer(astype=int, 
    columns=column_locs(df, ['home_code', 'purpose_code', ]))

scalars = Pipeline([
    ('inputs', FeatureUnion([('text', text_model), 
                             ('attrs', linear_vars)])),
    ('scaler', StandardScaler(with_mean=False)),
    ('clf', LogisticRegression()),
])

categories = Pipeline([
    ('inputs', category_vars),
    ('clf', RandomForestClassifier()),
])

union = Pipeline([
    ('inputs', FeatureUnion([('scalars', scalars), 
                             ('categories', categories)
                             ])),
    ('clf', LogisticRegression(C=.01, penalty='l1')),
])

paramgrid = {
    'inputs__scalars__clf__penalty': ('l1', 'l2'),
    'inputs__scalars__clf__C': (.001, .01, .1),#, 1., 10.),
    'inputs__categories__clf__n_estimators': (1000,),
}

grid_search = GridSearchCV(union, paramgrid, n_jobs=-1, verbose=1, cv=5)
grid_search.fit(df.values, ~df.failed.values)
display(grid_search.best_score_)
```

### Model Evaluation

```python
df_test = select_data(datb, balance=True)
X, y = featurize(df_test, reuse=True)

pred = grid_search.best_estimator_.predict(X)
df_test['predict'] = pred.astype(bool)
# df.groupby('subgrade_code').apply(lambda x: 
#                                   (mean(~x.target & x.predict), 
#                                    mean(x.target & ~x.predict), 
#                                    len(x.index)))
# df.groupby('subgrade_code').apply(lambda x: 
#                                   (x.target[x.predict].mean(), 
#                                    x.predict[x.target].mean()))
```
