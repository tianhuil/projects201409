```python


playlist_randomrandom_name = 'test_random.csv'
playlist_random_name = 'test_nodirection.csv'
playlist_year_tight_name = 'test_year_tight.csv'
playlist_year_loose_name = 'test_year_loose.csv'
playlist_artfam_tight_name = 'test_artfam_tight.csv'
playlist_artfam_loose_name = 'test_artfam_loose.csv'
playlist_tempo_tight_name = 'test_tempo_tight.csv'
playlist_tempo_loose_name = 'test_tempo_loose.csv'




# load playlists

pl_random = pd.read_csv(playlist_randomrandom_name,header=0)
pl_nodir = pd.read_csv(playlist_random_name,header=0)
pl_year_tight = pd.read_csv(playlist_year_tight_name,header=0)
pl_year_loose = pd.read_csv(playlist_year_loose_name,header=0)
pl_artfam_tight = pd.read_csv(playlist_artfam_tight_name,header=0)
pl_artfam_loose = pd.read_csv(playlist_artfam_loose_name,header=0)
pl_tempo_tight = pd.read_csv(playlist_tempo_tight_name,header=0)
pl_tempo_loose = pd.read_csv(playlist_tempo_loose_name,header=0)

#for i in range(pl_nodir.shape[1]):
#    pl_nodir.columns.values[i] = (pl_nodir.columns.values[i][1:])

    
print [x for x in list(pl_nodir.columns.values)]

    
#print (pl_nodir.columns.values[3])


#print pl_nodir.values[150][3]

year_random      = np.array([x[3] for x in pl_random.values])
year_nodir      = np.array([x[3] for x in pl_nodir.values])
year_year_tight = np.array([x[3] for x in pl_year_tight.values])
year_year_loose = np.array([x[3] for x in pl_year_loose.values])

artfam_random        = np.array([float(x[23][1:-1]) for x in pl_random.values])
artfam_nodir        = np.array([float(x[23][1:-1]) for x in pl_nodir.values])
artfam_artfam_tight = np.array([float(x[23][1:-1]) for x in pl_artfam_tight.values])
artfam_artfam_loose = np.array([float(x[23][1:-1]) for x in pl_artfam_loose.values])

tempo_random        = np.array([float(x[13]) for x in pl_random.values])
tempo_nodir        = np.array([(x[13]) for x in pl_nodir.values])
tempo_tempo_tight = np.array([(x[13]) for x in pl_tempo_tight.values])
tempo_tempo_loose = np.array([(x[13]) for x in pl_tempo_loose.values])


#print len(year_nodir), len(year_year_loose), len(year_year_tight)

# draw 
print "Year"
plt.plot(range(0,  200), year_random,'.k', label='random')
plt.plot(range(250,450), year_nodir,'.b', label='graph, no orientation')
plt.plot(range(500,700), year_year_loose,'.r', label='graph, year orientation')
plt.plot(range(750,950), year_year_tight,'.g', label='graph, year orientation tight')
plt.legend(loc='lower right')
plt.ylabel('year')
plt.show()

print "Year jumps"
plt.plot(range(0,  200-1),     year_random[1:] - year_random[:-1],'.k', label='random')
plt.plot(range(250,450-1),     year_nodir[1:] - year_nodir[:-1],'.b', label='graph, no orientation')
plt.plot(range(500,700-1), year_year_loose[1:] - year_year_loose[:-1],'.r', label='graph, year orientation')
plt.plot(range(750,950-1), year_year_tight[1:] - year_year_tight[:-1],'.g', label='graph, year orientation tight')
plt.legend(loc='lower right')
plt.ylabel('year jump')
plt.show()

print "Artist familiarity"
plt.plot(range(0,  200), artfam_random,'.k', label='random')
plt.plot(range(250,450), artfam_nodir,'.b', label='graph, no orientation')
plt.plot(range(500,700), artfam_artfam_loose,'.r', label='graph, year orientation')
plt.plot(range(750,950), artfam_artfam_tight,'.g', label='graph, year orientation tight')
plt.legend(loc='lower right')
plt.ylabel('familiarity')
plt.show()

print "Artist familiarity jumps"
plt.plot(range(0,  200-1),     artfam_random[1:] - artfam_random[:-1],'.k', label='random')
plt.plot(range(250,450-1),     artfam_nodir[1:] - artfam_nodir[:-1],'.b', label='graph, no orientation')
plt.plot(range(500,700-1), artfam_artfam_loose[1:] - artfam_artfam_loose[:-1],'.r', label='graph, year orientation')
plt.plot(range(750,950-1), artfam_artfam_tight[1:] - artfam_artfam_tight[:-1],'.g', label='graph, year orientation tight')
plt.legend(loc='lower right')
plt.ylabel('familiarity jump')
plt.show()



print "Tempo"
plt.plot(range(0,  200), tempo_random,'.k', label='random')
plt.plot(range(250,450), tempo_nodir,'.b', label='graph, no orientation')
plt.plot(range(500,700), tempo_tempo_loose,'.r', label='graph, year orientation')
plt.plot(range(750,950), tempo_tempo_tight,'.g', label='graph, year orientation tight')
plt.legend(loc='upper right')
plt.ylabel('tempo')
plt.ylim(60,300)

plt.show()

print "Tempo jumps"
plt.plot(range(0,  200-1),     tempo_random[1:] - tempo_random[:-1],'.k', label='random')
plt.plot(range(250,450-1),     tempo_nodir[1:] - tempo_nodir[:-1],'.b', label='graph, no orientation')
plt.plot(range(500,700-1), tempo_tempo_loose[1:] - tempo_tempo_loose[:-1],'.r', label='graph, year orientation')
plt.plot(range(750,950-1), tempo_tempo_tight[1:] - tempo_tempo_tight[:-1],'.g', label='graph, year orientation tight')
plt.legend(loc='upper right')
plt.ylabel('tempo jump')
plt.ylim(-150,200)

plt.show()
```
