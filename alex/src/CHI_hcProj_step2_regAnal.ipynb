```python
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib as plt

from sklearn import datasets
from sklearn import linear_model
from sklearn import utils
from sklearn import cross_validation
from scipy import stats # for Pearson correlations
```

## Transforming Data to Better Tell a Story ##

```python
# Load df
df = pd.read_csv('../data/mergedData.csv')

# *** Convert Numbers to make them more interpretable:***

# calculate pop density in people / sqmile
df['popDen2010'] = df.comPop2010.div(df.area)*5280**2 # people per sqr mile

# scale walkscore.com scores: All scored represent perentage score wrt rest of city
def rescaleWalkScore(s):
    return 100*(s-min(s))/(max(s)-min(s))

df.walkScore = rescaleWalkScore(df.walkScore)
df.tranScore = rescaleWalkScore(df.tranScore)
df.bikeScore = rescaleWalkScore(df.bikeScore)

# normalize crime in community by population (crimes communited per 100,000 people within the community, per year)
cols = [col.find('_TA')!=-1 for col in df.columns]
colsNames = []
for col in df.columns:
    if col.find('_TA')!=-1:
        colsNames.append(col)
df2 = df.ix[:,cols].div(df.comPop2010, axis='index') # number of crimes per person, that year
df2.columns = df2.columns+'_per'

df = pd.concat([df, df2], axis = 1).drop(colsNames,axis=1)
```

# clean up all string columns from df
def rem_str_cols(df):
    colNamesToDrop = []
    for colName in df.columns:
        if type(df[colName].ix[1]) == type(''):
            colNamesToDrop.append(colName)
    return df.drop(colNamesToDrop,axis=1)
