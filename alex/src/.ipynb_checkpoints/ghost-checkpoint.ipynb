<b>Deliverable:</b> I want an algorithm that determines the extent that local crime rate affects housing price by housing type (htpye) with respect to the city wide average housing price by htype.


<b>Approach:</b> The most basic approach would be to characterize the correlation between absolute house price and crime.  The problem with finding one global correlation curve between these two variables is that other factors such as proximity to city center, commute time, etc. are not factored in.
<P>
Instead, I will take a localized approach, something like this (using Chicago Data):
<ol>
<li> Discretize Chicago into regions containing 8-10 communities (77 in total)
<li> Calculate a characteristic curve: regional housing price vs. crimerate (linear regression)
<li> With the characteristic curve, estimate the localized effect of crime on housing price
</ol>


Sample Data:

```python
from pandas import *
```

```python
cols = ['rgn','communCode','crime13','price13']
numComms = 77
crime13 = rand(numComms)
price13 = 100000. - 100000.*crime13

def rgnAry(n):
    out = zeros(n)
    tick = 1
    for idx in range(n):
        out[idx] = tick
        if (idx+1) % 8 == 0:
            tick += 1
    return out
    

d = {cols[0]: Series(rgnAry(numComms)), cols[1]: Series(arange(1,numComms+1)), cols[2]: Series(crime13), cols[3]: Series(price13)}
```

```python
df = DataFrame(d)
df
```
