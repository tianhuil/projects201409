# The Goal of the Following:

* Create arrays of crime data and housing prices, measure $p(\tau,\Delta |$ chg in crime$)$, the PDF for a percent change in housing price ($\Delta$) occuring at a time lag ($\tau$, before or after) a significant change in crime rate. 
* Graphs of $\langle \Delta \rangle (\tau)$ would be useful

```python
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pylab as plt
```

```python
city = 'Chicago'
cty2BmedSale = pd.read_csv('../data/City_Zhvi_2bedroom.csv') # Zillow 2B data by city
neib2BmedSale = pd.read_csv('../data/Neighborhood_Zhvi_2bedroom.csv') # Zillow 2B data by neighbourhood
df_neib2comm = pd.read_csv('../data/neibCom_zillowData.csv') # CHI Neib --> Community dictionary

# Filtering out Chicago Data from other data
cty2BmedSale = cty2BmedSale[cty2BmedSale.RegionName == city]
cty2BmedSale = cty2BmedSale.ix[:,'1996-04':'2014-07'].T
cty2BmedSale.columns = ['salePrice']

# 2B prices over time for each neighbourhood (neib)
neib2BmedSale = neib2BmedSale[neib2BmedSale.City == city]
df_prices = neib2BmedSale.drop(['Metro','City','State','CountyName'],axis=1)
df_prices = df_prices.rename(columns={'RegionName':'neib'}).set_index('neib')
df_prices = df_prices.T # Transposes, so I can more easily do time series analysis
df_prices.index = pd.to_datetime(df_prices.index) # Changes indexes to a datetime object
df_prices['CityMed'] = cty2BmedSale.salePrice # adds city wide med saleprice
df_prices['Year'] = [df_prices.index[k].year for k in range(len(df_prices))] # can be more efficient
df_prices['Month'] = [df_prices.index[k].month for k in range(len(df_prices))]

# Chi price index (wrt city-wide median price)
df_priceIdxs = df_prices.div(df_prices.CityMed, axis='index')
df_priceIdxs['Year'] = df_prices.Year
df_priceIdxs['Month'] = df_prices.Month
df_priceIdxs['CityMed'] = df_prices.CityMed # Keeps track of current price
```

```python
df_priceIdxsT = df_priceIdxs.T.reset_index()
df_priceIdxsT = pd.merge(df_neib2comm,df_priceIdxsT, on='neib', how ='right') # I have checked, and yes, all communities are accounted for in this merge
df_comNumPriceIdxsT = df_priceIdxsT[:72].groupby('comNum').mean()
cty2BmedSale = df_priceIdxsT.iloc[72,3:]

# Now we have city medians, and community prices normalized by the city median
# I now want to get crime graphs
```
